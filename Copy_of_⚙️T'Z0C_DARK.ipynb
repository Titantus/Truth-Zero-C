````python
# @title
import numpy as np
import matplotlib.pyplot as plt

# --- Implement calculate_nu_eff_profile function ---

def calculate_nu_eff_profile(r_values, rho_nu_profile, kernel_id='tetra_lock_gaussian', sigma_kpc=2.0):
    """
    Calculates the radial nu_eff (effective viscosity) profile for a galaxy.
    r_values: numpy array of radial distances (e.g., in kpc).
    rho_nu_profile: numpy array of rho_nu (neutrino exhaust density) at each r_value.
    kernel_id: Identifier for the geometric kernel (e.g., 'tetra_lock_gaussian').
    sigma_kpc: Characteristic width for the Gaussian kernel in kpc.
    """
    nu_eff_profile = np.zeros_like(r_values, dtype=float)
    dr = np.diff(r_values)[0] # Assuming uniform spacing for integration

    # Define the geometric kernel Kgeom
    def tetra_lock_gaussian_kernel(r_prime, r_center, sigma):
        # K(r,r') is function of |r-r'| for radial symmetry, hence r_prime is a source at r_center
        return np.exp(-((r_prime - r_center)**2) / (2 * sigma**2))

    for i, r_center in enumerate(r_values):
        # For each r_center, calculate the kernel's influence across all r_values
        if kernel_id == 'tetra_lock_gaussian':
            # Kgeom(r_prime, r_center) - this is Kgeom(|r_center - r_prime|)
            kernel_values = tetra_lock_gaussian_kernel(r_values, r_center, sigma_kpc)
        else:
            raise ValueError(f"Unknown kernel_id: {kernel_id}")

        # Integrate rho_nu(r') * Kgeom(|r_center - r_prime|) over all r_prime
        # Use numerical integration (trapezoidal rule or sum for discrete points)
        # Note: The integral in Appendix D is d3r', implying a volume integral.
        # For a 1D radial profile, this simplifies to integrating over r' with r'^2 factor for spherical coords.
        # However, for simplicity and direct mapping to the given formula's notation ∫ρν(r′)Kgeom(|r−r′|)d3r′,
        # and assuming Kgeom already incorporates dimensionality if needed, we'll do a direct 1D integral here.
        # The factor 4*pi*r_values**2 should ideally be part of d3r' for spherical symmetry if Kgeom is only radial.
        # For this simplified model, we will use a simple 1D summation proportional to dr.
        integrand = rho_nu_profile * kernel_values
        nu_eff_profile[i] = np.trapezoid(integrand, r_values) # Numerical integration

        # We need a proportionality constant as per νeff(r)∝∫ρν(r′)Kgeom(|r−r′|)d3r′
        # For now, let's keep it implicitly in the units, or assume a scaling factor.
        # A constant of proportionality (e.g., k_nu_eff_scaling) can be introduced here if needed.

    return nu_eff_profile

# --- Call the calculate_nu_eff_profile function ---

# Assuming radial_distances and rho_nu_galactic_profile are available from the previous step
# central_baryonic_number_density is 1e3 and R_d is 5

# Need to re-run the previous rho_nu calculation to ensure continuity if notebook was reset
# (This code block assumes the state from previous executed cells is available)

# Let's re-define baryonic_density_profile and elements_tz0c_params to be safe
def baryonic_density_profile(r, rho_0=1e3, R_d=5):
    return rho_0 * np.exp(-r / R_d)

# Define key elements and their approximate solar (relative) abundances by number.
galactic_elements_abundances = {
    'H': 0.92, 'He': 0.07, 'C': 0.001, 'O': 0.005, 'Fe': 0.0005
}

total_abundance = sum(galactic_elements_abundances.values())
normalized_abundances = {k: v / total_abundance for k, v in galactic_elements_abundances.items()}

elements_tz0c_params = {}
for symbol, abundance in normalized_abundances.items():
    el_data = registry.get(symbol)
    if el_data:
        elements_tz0c_params[symbol] = {
            'abundance': abundance,
            'conflict_index_omega': el_data.get('conflict_index_omega', 0.0),
            'partition_f_nu': el_data.get('partition_f_nu', 0.001),
            'rho_nu_conflict_coupling': el_data.get('rho_nu_conflict_coupling', 1.25e-4)
        }
    # else: print(f"Warning: Element {symbol} not found in registry. Skipping.")

radial_distances = np.linspace(0.1, 25, 100) # kpc
rho_nu_galactic_profile = calculate_rho_nu_profile(
    radial_distances,
    lambda r: baryonic_density_profile(r, rho_0=1e3, R_d=5),
    elements_tz0c_params
)

# Now, calculate the nu_eff profile
nu_eff_galactic_profile = calculate_nu_eff_profile(radial_distances, rho_nu_galactic_profile)

# --- Plot the calculated nu_eff_galactic_profile ---
plt.figure(figsize=(10, 6))
plt.plot(radial_distances, nu_eff_galactic_profile, color='lightcoral', linewidth=2)
plt.title('Radial Effective Viscosity (ν_eff) Profile in a T\'Z0C Galaxy', fontsize=14)
plt.xlabel('Radial Distance (kpc)', fontsize=12)
plt.ylabel('ν_eff (Arbitrary Units)', fontsize=12)
plt.yscale('log') # Use logarithmic scale for y-axis
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("Radial \u03bd_eff (effective viscosity) profile calculated and visualized.")
````

````python
# @title
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define baryonic_velocity_curve function ---

def baryonic_velocity_curve(r, M_baryonic, R_d):
    """
    Calculates the rotational velocity due to baryonic matter (thin disk approximation).
    r: radial distance (kpc).
    M_baryonic: total baryonic mass (M_solar).
    R_d: disk scale length (kpc).

    A common and simplified empirical form for the velocity from an exponential disk is used here,
    based on the enclosed mass within a given radius.
    """
    G_const = 4.3009e-6 # Gravitational constant in (kpc/M_solar) * (km/s)^2

    # Simplified enclosed mass for an exponential disk, M(r) = M_baryonic * (1 - (1 + r/R_d) * exp(-r/R_d))
    # This formula approximates the cumulative mass distribution for an exponential disk.
    M_enclosed_r = M_baryonic * (1 - (1 + r/R_d) * np.exp(-r/R_d))

    # Avoid division by zero at r=0, and ensure physically reasonable behavior
    if r == 0:
        return 0.0

    # V_circ^2 = G * M_enclosed(r) / r
    V_squared = G_const * M_enclosed_r / r

    # Ensure non-negative under square root, as M_enclosed_r should always be non-negative.
    return np.sqrt(np.maximum(0, V_squared))

print("baryonic_velocity_curve function defined.")
````

````python
# @title
import numpy as np

# --- 2. Define viscous_velocity_contribution function ---

def viscous_velocity_contribution(r, nu_eff_profile, radial_distances, nu_eff_scaling_factor):
    """
    Calculates the velocity contribution from the effective viscosity (the 'dark matter effect').
    r: radial distance (kpc).
    nu_eff_profile: numpy array of nu_eff at each radial_distances point.
    radial_distances: numpy array of radial distances corresponding to nu_eff_profile.
    nu_eff_scaling_factor: a scaling factor to adjust the magnitude of the viscous contribution.

    Based on T'Z0C theory, this contribution should increase with nu_eff.
    A suitable mathematical relationship can be proportional to the square root of nu_eff,
    as velocity is often related to the square root of a force or potential.
    We need to interpolate nu_eff_profile to get nu_eff at a specific 'r'.
    """
    # Interpolate nu_eff_profile to get nu_eff at the given 'r'
    # np.interp requires sorted x-coordinates, which radial_distances should be.
    if r < radial_distances[0] or r > radial_distances[-1]:
        # Handle out-of-bounds by returning 0 or extending the closest value
        # For simplicity, we'll return 0 if outside the defined range
        return 0.0

    nu_eff_at_r = np.interp(r, radial_distances, nu_eff_profile)

    # The contribution is proportional to the square root of nu_eff, scaled by nu_eff_scaling_factor.
    # This models the 'viscous drag' effect proposed by T'Z0C.
    # Ensure nu_eff_at_r is non-negative before taking the square root.
    return nu_eff_scaling_factor * np.sqrt(np.maximum(0, nu_eff_at_r))

print("viscous_velocity_contribution function defined.")
````

````python
# @title
import numpy as np

# --- 3. Define total_rotational_velocity function ---

def total_rotational_velocity(r, baryonic_mass_profile_func, nu_eff_profile_func, radial_distances_for_nu_eff, M_baryonic, R_d, nu_eff_scaling_factor):
    """
    Calculates the total rotational velocity curve for a galaxy by combining baryonic and viscous contributions.
    r: radial distance (kpc).
    baryonic_mass_profile_func: function to calculate velocity due to baryonic matter.
    nu_eff_profile_func: function to calculate velocity contribution from effective viscosity.
    radial_distances_for_nu_eff: radial distances array used for nu_eff_profile_func interpolation.
    M_baryonic: total baryonic mass (M_solar).
    R_d: disk scale length (kpc).
    nu_eff_scaling_factor: scaling factor for the viscous contribution.

    Combines contributions in quadrature: V_rot = sqrt(V_baryonic^2 + V_viscous^2).
    """
    # Calculate baryonic velocity contribution
    v_baryonic = baryonic_mass_profile_func(r, M_baryonic, R_d)

    # Calculate viscous velocity contribution
    v_viscous = nu_eff_profile_func(r, nu_eff_galactic_profile, radial_distances_for_nu_eff, nu_eff_scaling_factor)

    # Combine in quadrature
    v_rot_squared = v_baryonic**2 + v_viscous**2

    return np.sqrt(v_rot_squared)

print("total_rotational_velocity function defined.")
````
