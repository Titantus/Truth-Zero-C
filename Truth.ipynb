```python
# @title
import re

# ==========================================
# UNIFIED T0C EPISTEMIC PARSER
# ==========================================

# 1. Define T0C Truth-Class logic triggers
truth_classes = {
    "⧉ Observable": {
        "symbol": "⧉",
        "weight": 4,
        "keywords": ['measured', 'observed', 'detected', 'empirical', 'data shows', 'confirmed', 'seen', 'verified', 'astronomical observations', 'atomic clocks', 'LIGO/Virgo', 'experiment', 'laboratory']
    },
    "⧠ Model-Established": {
        "symbol": "⧠",
        "weight": 3,
        "keywords": ['standard model predicts', 'law of', 'consensus', 'theory', 'model', 'framework', 'formulated', 'proposes', 'geodesics', 'equivalence principle', 'cosmology', 'big bang model', 'equations', 'established', 'calculates', 'relativity', 'quantum mechanics', 'symmetries', 'particle frameworks']
    },
    "⇢ Derived Prediction": {
        "symbol": "⇢",
        "weight": 2,
        "keywords": ['mathematically implies', 'extrapolating from', 'should yield', 'predicts', 'derived from', 'existence of', 'gravitational waves', 'black holes', 'singularities', 'requires', 'propose', 'dictates', 'emerges']
    },
    "→ ? Potential (Unverified)": {
        "symbol": "→ ?",
        "weight": 1,
        "keywords": ['assume', 'if we posit', r'in a \d+-dimensional space', 'beyond the observable', 'speculative', 'highly speculative', 'theoretical', 'wormholes', 'limits of the theory', 'new physics is needed', 'posit', 'extra dimensions', 'string', 'planck length', 'unobservable', 'multiverse', 'loop quantum', 'spin network', 'unobserved']
    }
}

def parse_epistemic_text(text):
    text = text.replace('\n', ' ')
    sentences = re.split(r'(?<=[.!?]) +', text.strip())
    results = []
    stats = {"⧉": 0, "⧠": 0, "⇢": 0, "→ ?": 0} # Initialize for all symbols

    for sentence in sentences:
        if not sentence.strip():  # Skip empty sentences after strip
            continue

        assigned_symbol = "⧠"  # Default to established model logic unless overwritten
        lowest_weight = 3       # Corresponds to ⧠ Model-Established

        lower_sentence = sentence.lower()

        # First pass: find the most speculative (lowest weight) category
        for t_class_name, data in truth_classes.items():
            for keyword in data["keywords"]:
                # Handle regex keywords
                if data["symbol"] == "→ ?" and keyword.startswith(r'in a \d+-dimensional space'):
                    if re.search(keyword, lower_sentence):
                        if data["weight"] < lowest_weight:
                            assigned_symbol = data["symbol"]
                            lowest_weight = data["weight"]
                elif keyword in lower_sentence:
                    if data["weight"] < lowest_weight:
                        assigned_symbol = data["symbol"]
                        lowest_weight = data["weight"]

        # Second pass: If an observable keyword is present, it overrides to ⧉
        for keyword in truth_classes["⧉ Observable"]["keywords"]:
            if keyword in lower_sentence:
                assigned_symbol = "⧉"
                break # No need to check other observable keywords

        results.append((assigned_symbol, sentence))
        # Ensure the symbol is one of the keys in stats
        if assigned_symbol not in stats:
            # This should ideally not happen if symbols are consistent, but good for robustness
            stats[assigned_symbol] = 0
        stats[assigned_symbol] += 1

    return results, stats
```

```python
# @title
import os
import urllib.request
import xml.etree.ElementTree as ET
import urllib.parse
import re

# 2. Define search terms for theories
search_terms = {
    'String Theory': 'string theory',
    'Loop Quantum Gravity': 'loop quantum gravity',
    'Quantum Field Theory': 'quantum field theory'
}

# 3. Specify base directory and create it if it doesn't exist
base_dir = 'arxiv_papers'
os.makedirs(base_dir, exist_ok=True)

# 4. Initialize an empty dictionary to store metadata
all_downloaded_papers = {}

print(f"Downloading papers to: {os.path.abspath(base_dir)}\n")

# 5. Loop through each theory and its search query
for theory_name, query in search_terms.items():
    print(f"\n--- Processing {theory_name} ---")

    # a. Create a subdirectory for the current theory
    theory_dir = os.path.join(base_dir, theory_name.replace(' ', '_'))
    os.makedirs(theory_dir, exist_ok=True)

    # b. Construct the arXiv API query URL
    # Requesting a maximum of 50 results, sorted by relevance
    encoded_query = urllib.parse.quote(query)
    arxiv_api_url = f'http://export.arxiv.org/api/query?search_query=all:{encoded_query}&start=0&max_results=50&sortBy=relevance&sortOrder=descending'

    try:
        # c. Send an HTTP GET request to the arXiv API URL
        req_api = urllib.request.Request(arxiv_api_url, headers={'User-Agent': 'Mozilla/5.0'})
        with urllib.request.urlopen(req_api) as response:
            xml_data = response.read()

        # d. Parse the XML response
        root = ET.fromstring(xml_data)

        # Namespace for XML parsing
        namespace = {'atom': 'http://www.w3.org/2005/Atom'}

        download_count = 0
        # e. Iterate through each entry in the XML response
        for entry in root.findall('atom:entry', namespace):
            title_element = entry.find('atom:title', namespace)
            title = title_element.text.strip() if title_element is not None else "Untitled"
            pdf_url_element = entry.find('atom:link[@title="pdf"]', namespace)

            # Extract arXiv ID for robust deduplication and logging
            arxiv_id_element = entry.find('atom:id', namespace)
            # Extract only the ID part from the full URL string
            arxiv_id = arxiv_id_element.text.split('/')[-1] if arxiv_id_element is not None else title # Fallback to title if ID missing

            # Check if paper has already been downloaded (robust deduplication using arxiv_id)
            if arxiv_id in all_downloaded_papers:
                # print(f"  Skipping duplicate (already tracked): '{title}' (ID: {arxiv_id})") # Commented to reduce verbosity
                continue

            if pdf_url_element is not None:
                pdf_url = pdf_url_element.attrib['href']

                # ii. Construct a safe filename for the PDF
                # Replace invalid characters with underscores
                safe_title = re.sub(r'[^a-zA-Z0-9_.-]', '_', title)
                filename = f"{arxiv_id}_{safe_title}.pdf" # Use arXiv ID to prevent collisions

                # iii. Define the full path where the PDF will be saved
                file_path = os.path.join(theory_dir, filename)

                try:
                    # iv. Download the PDF
                    # Add a User-Agent header to mimic a browser, which can help prevent 403 errors
                    req_pdf = urllib.request.Request(pdf_url, headers={'User-Agent': 'Mozilla/5.0'})
                    with urllib.request.urlopen(req_pdf) as pdf_response:
                        with open(file_path, 'wb') as outfile:
                            outfile.write(pdf_response.read())

                    # v. Store metadata
                    all_downloaded_papers[arxiv_id] = { # Use arxiv_id as key for unique storage
                        'title': title,
                        'theory': theory_name,
                        'file_path': file_path
                    }
                    download_count += 1
                    # vi. Print confirmation message (reduced verbosity)
                    # print(f"  Downloaded: '{title}' (ID: {arxiv_id}) for {theory_name}")
                except Exception as e:
                    print(f"  Error downloading '{title}' (ID: {arxiv_id}): {e}")
            else:
                # print(f"  No PDF link found for '{title}' (ID: {arxiv_id})") # Commented to reduce verbosity
                pass # Skip entries without PDF links without printing
        print(f"  Successfully downloaded {download_count} papers for {theory_name}.")

    except urllib.error.URLError as e:
        print(f"  Error fetching data for {theory_name} from arXiv API: {e}")
    except ET.ParseError as e:
        print(f"  Error parsing XML for {theory_name}: {e}")

# 6. Print summary
print("\n--- Download Summary ---")
for theory_name in search_terms.keys():
    count = sum(1 for paper_data in all_downloaded_papers.values() if paper_data['theory'] == theory_name)
    print(f"{theory_name}: {count} papers downloaded")

print(f"Total unique papers downloaded: {len(all_downloaded_papers)}")
```

```python
# @title
import PyPDF2 # Ensure PyPDF2 is imported
import io
import os

# Ensure `parse_epistemic_text` and `truth_classes` are available from previous cells
# If they were defined within a cell that's not always executed or is mutable, they should be redefined here for robustness.
# For this execution, we assume they are globally available after previous cell execution.

# Initialize a dictionary to store all heatmaps and parsed sentences
all_paper_epistemic_data = {}

print("\n--- Extracting Text and Parsing PDFs ---")
processed_count = 0
error_count = 0

for arxiv_id, paper_data in all_downloaded_papers.items():
    file_path = paper_data['file_path']
    theory_name = paper_data['theory']
    title = paper_data['title']

    if not os.path.exists(file_path):
        print(f"  Skipping '{title}' (ID: {arxiv_id}) - File not found.")
        error_count += 1
        continue

    try:
        # Open the PDF file in binary read mode
        with open(file_path, 'rb') as pdf_file_obj:
            reader = PyPDF2.PdfReader(pdf_file_obj)

            # Extract text from all pages
            full_text = " ".join([page.extract_text() for page in reader.pages if page.extract_text()])

            # Apply the epistemic parser
            parsed_sentences, heatmap_stats = parse_epistemic_text(full_text)

            # Store the results
            all_paper_epistemic_data[arxiv_id] = {
                'title': title,
                'theory': theory_name,
                'file_path': file_path,
                'heatmap': heatmap_stats,
                'parsed_sentences': parsed_sentences # Storing for potential detailed analysis later
            }
            processed_count += 1
            # print(f"  Processed: '{title}' (ID: {arxiv_id}) for {theory_name}")

    except PyPDF2.errors.PdfReadError as e:
        print(f"  Error reading PDF '{title}' (ID: {arxiv_id}): {e}")
        error_count += 1
    except Exception as e:
        print(f"  An unexpected error occurred for '{title}' (ID: {arxiv_id}): {e}")
        error_count += 1

print(f"\n--- PDF Processing Summary ---")
print(f"Total PDFs processed: {processed_count}")
print(f"Total PDFs with errors or skipped: {error_count}")
print(f"Total papers with epistemic data: {len(all_paper_epistemic_data)}")
```

```python
# @title
import pandas as pd

# Initialize dictionaries to store aggregated counts for each theory
theory_aggregated_stats = {
    'String Theory': {'⧉': 0, '⧠': 0, '⇢': 0, '→ ?': 0, 'total_sentences': 0},
    'Loop Quantum Gravity': {'⧉': 0, '⧠': 0, '⇢': 0, '→ ?': 0, 'total_sentences': 0},
    'Quantum Field Theory': {'⧉': 0, '⧠': 0, '⇢': 0, '→ ?': 0, 'total_sentences': 0}
}

print("\n--- Calculating Average Epistemic Heatmaps for Each Theory ---")

for arxiv_id, paper_data in all_paper_epistemic_data.items():
    theory = paper_data['theory']
    heatmap = paper_data['heatmap']
    parsed_sentences = paper_data['parsed_sentences']

    # Sum the counts for each truth class
    for symbol, count in heatmap.items():
        theory_aggregated_stats[theory][symbol] += count

    # Sum the total number of sentences for the paper
    theory_aggregated_stats[theory]['total_sentences'] += len(parsed_sentences)

# Calculate percentages for each theory
for theory_name, stats in theory_aggregated_stats.items():
    print(f"\n### {theory_name} Average Epistemic Heatmap ###")
    total_sentences_for_theory = stats['total_sentences']

    if total_sentences_for_theory > 0:
        for symbol in ['⧉', '⧠', '⇢', '→ ?']:
            percentage = (stats[symbol] / total_sentences_for_theory) * 100
            print(f"[{symbol}]: {percentage:.2f}%")
    else:
        print("No sentences processed for this theory.")
```
