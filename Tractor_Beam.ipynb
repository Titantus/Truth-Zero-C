```python
# @title
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import copy # Import copy module for deepcopy

# --- T'Z0C: Instructional Battery (Vortex Wind-up) ---

# Set a dark background style for plots
plt.style.use('dark_background')

# Global Parameters for the simulation
num_particles = 1000  # Number of particles in the simulation
PARTICLE_MASS_PER_PARTICLE_KG = 0.01 # kg, mass of each particle
CRITICAL_VELOCITY_THRESHOLD_MPS = 20.0 # m/s, beyond this speed, a particle incurs full clutch-tax (heat)
spin_rate = 0.1 # Rate at which coherence builds
chaos_level = 0.05 # Initial level of environmental chaos

# Bounce-Gap Parameters (Laminar Maximum) - now global
BOUNCE_GAP_LOWER_MPS = 500.0  # m/s, velocities below this are always successful
BOUNCE_GAP_UPPER_MPS = 550.0  # m/s, velocities above this are always failed (full clutch-tax)

# Function to define a Zero-Manifold Anchor
def define_zero_manifold_anchor(baseline_temp, delta_temp, position=(0, 0, 0)):
    return {
        'position': np.array(position),
        'temperature': baseline_temp - delta_temp,
        'asymmetry_factor': 1.0,
        'routing_mode': 'STRAIGHT'
    }

# Placeholder for apply_vector_contagion function
def apply_vector_contagion(particles, contagion_factor):
    # Simulate a simple alignment towards the average direction of neighbors
    # For this simplified version, let's just slightly bias their velocity towards the release direction
    # This function is a placeholder and can be made more sophisticated.

    # Define a target direction (e.g., the overall average alignment if available, or a fixed direction)
    # For now, let's assume a general 'forward' direction, or no specific contagion effect other than initial alignment.
    # In a full simulation, this would involve local interactions and alignment.
    # For the scope of the sweep and release pulse, we assume initial alignment is set.
    return particles

# Placeholder for apply_sheepdog_correction function
def apply_sheepdog_correction(particles, correction_strength, boundary_min, boundary_max):
    # This function would typically correct the positions/velocities of non-sheepdog particles
    # based on the behavior of sheepdog particles to maintain group cohesion or boundary.
    # For the current `execute_release_pulse` context, it's not directly applied as it's a one-shot event.
    # It's more relevant for continuous simulation (like in the sheepdog sweep).
    return particles

def calculate_lattice_tension(particles):
    # Calculate an approximate lattice tension (e.g., variance in particle positions)
    # Lower variance indicates higher tension/cohesion
    if len(particles['position']) < 2: # Handle case with too few particles
        return 0.0

    # Centroid of the particle cloud
    centroid = np.mean(particles['position'], axis=0)

    # Distance of each particle from the centroid
    distances = np.linalg.norm(particles['position'] - centroid, axis=1)

    # Variance of these distances as a proxy for 'tension'
    # Smaller variance means particles are closer to each other, implying higher tension
    return np.var(distances)

def calculate_coherence(particles):
    # Calculate a simple measure of coherence based on alignment vectors
    # This is a placeholder and can be made more sophisticated if alignment vectors are dynamic
    if 'alignment_vector' not in particles or len(particles['alignment_vector']) == 0:
        return 0.0

    # Assuming alignment_vector is towards a general 'goal' direction (e.g. release_direction)
    # If it's a measure of internal alignment, we would compare vectors against each other.

    # For now, a placeholder: high value if all particles are 'instructed'
    if 'is_instructed' in particles:
        return np.sum(particles['is_instructed']) / len(particles['is_instructed']) # Fraction of instructed particles
    return 1.0 # Default if no specific instruction status

def simulate_vortex_windup(pulses, spin_rate, environment_noise):
    """
    Simulates how many "kicks" it takes to reach a coherent whirlwind state.
    """
    time = np.arange(0, pulses, 1)
    # Coherence builds exponentially as orientation overcomes chaos
    coherence = 1 - np.exp(-time * (spin_rate / environment_noise))
    return time, coherence

def calculate_release_force(peak_coherence, internal_energy):
    """
    Calculates the 'Newtonian Shot' when the vortex is released.
    """
    # The release force is the stored Asymmetry of the vortex
    return peak_coherence * internal_energy

def execute_release_pulse(particles_data, release_direction_vector, release_magnitude, critical_velocity_threshold, particle_mass, bounce_gap_lower, bounce_gap_upper):
    """
    Executes a single release pulse, applying a force to instructed particles
    and calculating effective KE and residue heat based on soft bounce-gap.

    Args:
        particles_data (dict): Dictionary containing 'position', 'velocity',
                               'alignment_vector', and 'is_instructed' arrays.
        release_direction_vector (np.array): Unit vector indicating the direction of the impulse.
        release_magnitude (float): The strength of the impulse to apply.
        critical_velocity_threshold (float): (Deprecated for soft bounce-gap, but kept for compatibility)
                                           Old threshold for clutch-tax failure.
        particle_mass (float): Mass of a single particle.
        bounce_gap_lower (float): Lower bound of the soft bounce-gap (velocities below this are always successful).
        bounce_gap_upper (float): Upper bound of the soft bounce-gap (velocities above this are always failed).
                                   Velocities within the gap fail probabilistically.

    Returns:
        tuple: (effective_kinetic_energy, residue_heat, successful_particles_count, failed_particles_count)
    """
    particles = particles_data
    num_particles = len(particles['position'])

    # Filter for instructed particles
    instructed_mask = particles['is_instructed'] == True

    if not np.any(instructed_mask):
        return 0.0, 0.0, 0, num_particles # No instructed particles, no effect

    # Calculate the potential impulse vector for instructed particles
    # The impulse is applied in the release_direction_vector scaled by release_magnitude
    potential_shot_impulse = particles['alignment_vector'][instructed_mask] * release_magnitude

    # Current velocities of instructed particles
    current_velocities_instructed = particles['velocity'][instructed_mask]

    # Calculate the potential new velocities if the impulse were perfectly applied
    potential_new_velocities_vector = current_velocities_instructed + potential_shot_impulse
    potential_new_velocities_magnitude = np.linalg.norm(potential_new_velocities_vector, axis=1)

    effective_kinetic_energy = 0.0
    residue_heat = 0.0
    successful_mask = np.zeros(num_particles, dtype=bool)
    failed_mask = np.zeros(num_particles, dtype=bool)

    # --- Soft Bounce-Gap Logic ---

    # Category 1: Particles entirely below the bounce-gap window are always successful
    successful_below_bounce_gap = potential_new_velocities_magnitude <= bounce_gap_lower
    successful_mask[np.where(instructed_mask)[0][successful_below_bounce_gap]] = True

    # Category 2: Particles entirely above the bounce-gap window are always failed
    failed_above_bounce_gap = potential_new_velocities_magnitude >= bounce_gap_upper
    failed_mask[np.where(instructed_mask)[0][failed_above_bounce_gap]] = True

    # Category 3: Particles within the bounce-gap window have a probabilistic failure
    in_bounce_gap_window = (potential_new_velocities_magnitude > bounce_gap_lower) & \
                           (potential_new_velocities_magnitude < bounce_gap_upper)

    velocities_in_bounce_gap = potential_new_velocities_magnitude[in_bounce_gap_window]

    if np.sum(in_bounce_gap_window) > 0:
        # Probability of failure increases linearly from 0% at lower bound to 100% at upper bound
        probability_of_failure = (velocities_in_bounce_gap - bounce_gap_lower) / \
                                 (bounce_gap_upper - bounce_gap_lower)

        random_draw = np.random.rand(np.sum(in_bounce_gap_window))
        probabilistic_failed_in_gap = random_draw < probability_of_failure

        # Map these back to the full instructed_mask indices
        instructed_indices_in_gap = np.where(instructed_mask)[0][in_bounce_gap_window]
        failed_mask[instructed_indices_in_gap[probabilistic_failed_in_gap]] = True
        successful_mask[instructed_indices_in_gap[~probabilistic_failed_in_gap]] = True

    # --- Apply outcomes ---

    # Calculate effective KE for successful particles
    successful_instructed_mask_indices = np.where(successful_mask & instructed_mask)[0]
    if len(successful_instructed_mask_indices) > 0:
        # Only count KE for particles that pass the bounce-gap successfully
        # We need to get their velocities *before* any potential clamping or resetting for failed particles
        final_successful_velocities_magnitude = potential_new_velocities_magnitude[successful_mask[instructed_mask]]
        effective_kinetic_energy = np.sum(0.5 * particle_mass * final_successful_velocities_magnitude**2)

    # Calculate residue heat for failed particles
    failed_instructed_mask_indices = np.where(failed_mask & instructed_mask)[0]
    if len(failed_instructed_mask_indices) > 0:
        # All kinetic energy of failed particles converts to residue heat
        failed_velocities_magnitude = potential_new_velocities_magnitude[failed_mask[instructed_mask]]
        residue_heat = np.sum(0.5 * particle_mass * failed_velocities_magnitude**2)

        # Reset velocity of failed particles to zero (they are 'dissipated' into heat)
        particles['velocity'][failed_instructed_mask_indices] = 0.0
        # Optionally, move failed particles to a 'residue' sink or remove them from consideration
        # For this simulation, we'll just zero their velocity.

    # Update velocities for successful particles based on the potential new velocities
    # Note: If no failures in the instructed_mask, all potential_new_velocities_vector become actual.
    # Otherwise, only the successful ones retain their calculated velocity.
    particles['velocity'][successful_instructed_mask_indices] = potential_new_velocities_vector[successful_mask[instructed_mask]]

    # Count successful and failed particles
    successful_particles_count = np.sum(successful_mask)
    failed_particles_count = np.sum(failed_mask) + (num_particles - np.sum(instructed_mask)) # Add uninstructed if they count as 'failed to align'

    # Ensure counts are within bounds
    successful_particles_count = min(successful_particles_count, num_particles)
    failed_particles_count = min(failed_particles_count, num_particles)

    # Adjust total counts for consistency (e.g., particles not instructed but also not failed)
    # For this function, let's focus on the instructed particles. Total instructed = successful + failed.
    total_instructed = np.sum(instructed_mask)
    successful_particles_count = np.sum(successful_mask[instructed_mask])
    failed_particles_count = np.sum(failed_mask[instructed_mask])

    return effective_kinetic_energy, residue_heat, successful_particles_count, failed_particles_count

# --- Vortex Wind-up Parameters (now using global params) ---
pulses = 100       # Number of instructional pulses

# --- Calculate Vortex Coherence and Release Power ---
t, coherence_data = simulate_vortex_windup(pulses, spin_rate, chaos_level)
release_power = calculate_release_force(coherence_data[-1], 500) # 500 units of internal heat

# --- Vortex Wind-up Visualization ---
plt.figure(figsize=(10, 5))
plt.plot(t, coherence_data, color='cyan', label='Instructional Coherence')
plt.axhline(y=1.0, color='r', linestyle='--', label='Universal Coherence (Limit)')
plt.title('Vortex Wind-up: Instructions turning Chaos into Structure')
plt.xlabel('Number of Pulses (Instructional Input)')
plt.ylabel('Coherence (%)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Vortex Wound to {coherence_data[-1]*100:.2f}% Coherence.")
print(f"Potential Release Force: {release_power:.2f} Newtonian Units.")


# --- Prepare a base 'instructed' particle state for deepcopy for the sweep ---
# This ensures each sweep iteration starts from the same initial coherent state.
base_particles_for_sweep = {}
base_particles_for_sweep['position'] = np.random.rand(num_particles, 3) * np.array([100, 100, 100]) # Example grid size
# Initialize velocities to something small and random
velocities_raw_base_sweep = np.random.uniform(-0.01, 0.01, size=(num_particles, 3))
velocities_magnitudes_base_sweep = np.linalg.norm(velocities_raw_base_sweep, axis=1)
velocities_unit_base_sweep = velocities_raw_base_sweep / (velocities_magnitudes_base_sweep[:, np.newaxis] + 1e-9) # Add small epsilon
base_particles_for_sweep['velocity'] = velocities_unit_base_sweep * np.random.uniform(0, 0.01, size=(num_particles, 1))

base_particles_for_sweep['torque'] = np.zeros((num_particles, 3)) # Not relevant for this specific test
release_direction = np.array([0, 0, 1]) # Consistent release direction
base_particles_for_sweep['alignment_vector'] = np.tile(release_direction, (num_particles, 1)) * 1.0 # Fully aligned
base_particles_for_sweep['is_instructed'] = np.full(num_particles, True, dtype=bool) # All particles are instructed

# Define the range of release_magnitude values to sweep
release_magnitudes_sweep = np.linspace(0, 1000, 100) # From 0 to 1000 with 100 points

# Lists to store results
effective_kinetic_energies = []
residue_heats = []
successful_particles_counts = []
failed_particles_counts = []

print(f"Starting sweep for {len(release_magnitudes_sweep)} release magnitudes...")

# Loop through each release magnitude
for i, rm in enumerate(release_magnitudes_sweep):
    # Create a deep copy of the base_particles_for_sweep for each iteration
    # This ensures that velocity changes from previous calls don't affect current call
    current_particles_for_sweep = copy.deepcopy(base_particles_for_sweep)

    # Call the execute_release_pulse function with new bounce-gap parameters
    effective_ke, residue_he, successful_count, failed_count = execute_release_pulse(
        current_particles_for_sweep,
        release_direction,
        rm,
        CRITICAL_VELOCITY_THRESHOLD_MPS, # Still pass for consistency, though bounce-gap is primary
        PARTICLE_MASS_PER_PARTICLE_KG,
        BOUNCE_GAP_LOWER_MPS,
        BOUNCE_GAP_UPPER_MPS
    )

    # Store the results
    effective_kinetic_energies.append(effective_ke)
    residue_heats.append(residue_he)
    successful_particles_counts.append(successful_count)
    failed_particles_counts.append(failed_count)

    if (i + 1) % 10 == 0:
        print(f"  Processed {i+1}/{len(release_magnitudes_sweep)} magnitudes.")

print("Sweep complete.")

# Identify the Laminar Maximum
max_ke_index = np.argmax(effective_kinetic_energies)
laminar_max_release_magnitude = release_magnitudes_sweep[max_ke_index]
max_effective_ke = effective_kinetic_energies[max_ke_index]
corresponding_residue_heat = residue_heats[max_ke_index]
corresponding_successful_count = successful_particles_counts[max_ke_index]
corresponding_failed_count = failed_particles_counts[max_ke_index]

print(f"\n--- Laminar Maximum Summary ---")
print(f"Optimal Release Magnitude (Laminar Maximum): {laminar_max_release_magnitude:.2f} units")
print(f"Max Effective Kinetic Energy: {max_effective_ke:.2e} Joules")
print(f"Corresponding Residue Heat: {corresponding_residue_heat:.2e} Joules")
print(f"Corresponding Successful Particles: {corresponding_successful_count}")
print(f"Corresponding Failed Particles: {corresponding_failed_count}")

# --- Plotting the results of the sweep ---
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Effective Kinetic Energy vs. Release Magnitude
axes[0, 0].plot(release_magnitudes_sweep, effective_kinetic_energies, color='blue', label='Effective Kinetic Energy')
axes[0, 0].set_title('Effective Kinetic Energy vs. Release Magnitude (Soft Bounce-Gap)')
axes[0, 0].set_xlabel('Release Magnitude (conceptual units)')
axes[0, 0].set_ylabel('Effective Kinetic Energy (Joules)')
axes[0, 0].axvline(x=laminar_max_release_magnitude, color='red', linestyle='--', label='Laminar Maximum')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Plot 2: Residue Heat vs. Release Magnitude
axes[0, 1].plot(release_magnitudes_sweep, residue_heats, color='orange', label='Residue Heat')
axes[0, 1].set_title('Residue Heat vs. Release Magnitude (Soft Bounce-Gap)')
axes[0, 1].set_xlabel('Release Magnitude (conceptual units)')
axes[0, 1].set_ylabel('Residue Heat (Joules)')
axes[0, 1].axvline(x=laminar_max_release_magnitude, color='red', linestyle='--', label='Laminar Maximum')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Plot 3: Number of Successful Particles vs. Release Magnitude
axes[1, 0].plot(release_magnitudes_sweep, successful_particles_counts, color='green', label='Successful Particles')
axes[1, 0].set_title('Successful Particles vs. Release Magnitude (Soft Bounce-Gap)')
axes[1, 0].set_xlabel('Release Magnitude (conceptual units)')
axes[1, 0].set_ylabel('Number of Particles')
axes[1, 0].axvline(x=laminar_max_release_magnitude, color='red', linestyle='--', label='Laminar Maximum')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Plot 4: Number of Failed Particles vs. Release Magnitude
axes[1, 1].plot(release_magnitudes_sweep, failed_particles_counts, color='red', label='Failed Particles')
axes[1, 1].set_title('Failed Particles vs. Release Magnitude (Soft Bounce-Gap)')
axes[1, 1].set_xlabel('Release Magnitude (conceptual units)')
axes[1, 1].set_ylabel('Number of Particles')
axes[1, 1].axvline(x=laminar_max_release_magnitude, color='red', linestyle='--', label='Laminar Maximum')
axes[1, 1].legend()
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

# --- 3D Visualization of Particle State at Laminar Maximum Release Magnitude ---
# Re-create a fresh set of base particles for visualization purposes
base_particles_for_viz = {}
base_particles_for_viz['position'] = np.random.rand(num_particles, 3) * np.array([100, 100, 100]) # Example grid size
# Initialize velocities to something small and random
velocities_raw_base_viz = np.random.uniform(-0.01, 0.01, size=(num_particles, 3))
velocities_magnitudes_base_viz = np.linalg.norm(velocities_raw_base_viz, axis=1)
velocities_unit_base_viz = velocities_raw_base_viz / (velocities_magnitudes_base_viz[:, np.newaxis] + 1e-9)
base_particles_for_viz['velocity'] = velocities_unit_base_viz * np.random.uniform(0, 0.01, size=(num_particles, 1))

base_particles_for_viz['torque'] = np.zeros((num_particles, 3))
# Ensure alignment vector is consistent with the release direction
base_particles_for_viz['alignment_vector'] = np.tile(release_direction, (num_particles, 1)) * 1.0
base_particles_for_viz['is_instructed'] = np.full(num_particles, True, dtype=bool)

# ADDITION: Include is_sheepdog attribute
num_sheepdogs_viz = int(num_particles * 0.1) # 10% sheep-dog particles
base_particles_for_viz['is_sheepdog'] = np.full(num_particles, False, dtype=bool)
sheepdog_indices_viz = np.random.choice(num_particles, num_sheepdogs_viz, replace=False)
base_particles_for_viz['is_sheepdog'][sheepdog_indices_viz] = True

# Run execute_release_pulse with this optimal magnitude for visualization particles
viz_particles = copy.deepcopy(base_particles_for_viz)

effective_ke_viz, residue_he_viz, successful_count_viz, failed_count_viz = execute_release_pulse(
    viz_particles,
    release_direction,
    laminar_max_release_magnitude, # CHANGED from release_power
    CRITICAL_VELOCITY_THRESHOLD_MPS,
    PARTICLE_MASS_PER_PARTICLE_KG,
    BOUNCE_GAP_LOWER_MPS,
    BOUNCE_GAP_UPPER_MPS
)

# To get the state *before* velocity reset for failed particles, we need to re-simulate.
# Or, given the counts, we can assume the first `successful_count_viz` particles were successful if they were sorted.
# For a more robust visualization without re-simulating the internal logic of execute_release_pulse:
# We need to compute potential_new_velocities_magnitude *outside* execute_release_pulse for fresh viz_particles.

# Let's re-calculate potential_new_velocities_magnitude for the viz_particles to accurately categorize them for plotting
# This logic is extracted from execute_release_pulse to categorize particles for viz.
instructed_mask_viz = viz_particles['is_instructed'] == True # All are instructed in this setup
potential_shot_impulse_viz = viz_particles['alignment_vector'][instructed_mask_viz] * laminar_max_release_magnitude # Use laminar_max_release_magnitude
current_velocities_instructed_viz = base_particles_for_viz['velocity'][instructed_mask_viz] # Use base_particles_for_viz for initial velocities
potential_new_velocities_vector_viz = current_velocities_instructed_viz + potential_shot_impulse_viz
potential_new_velocities_magnitude_viz = np.linalg.norm(potential_new_velocities_vector_viz, axis=1)

# --- Re-applying Soft Bounce-Gap Logic for Visualization ---
successful_mask_viz_temp = np.zeros(num_particles, dtype=bool)
failed_mask_viz_temp = np.zeros(num_particles, dtype=bool)

# Category 1: Particles entirely below the bounce-gap window are always successful
successful_below_bounce_gap_viz = potential_new_velocities_magnitude_viz <= BOUNCE_GAP_LOWER_MPS
successful_mask_viz_temp[successful_below_bounce_gap_viz] = True

# Category 2: Particles entirely above the bounce-gap window are always failed
failed_above_bounce_gap_viz = potential_new_velocities_magnitude_viz >= BOUNCE_GAP_UPPER_MPS
failed_mask_viz_temp[failed_above_bounce_gap_viz] = True

# Category 3: Particles within the bounce-gap window have a probabilistic failure
in_bounce_gap_window_viz = (potential_new_velocities_magnitude_viz > BOUNCE_GAP_LOWER_MPS) & \
                           (potential_new_velocities_magnitude_viz < BOUNCE_GAP_UPPER_MPS)

velocities_in_bounce_gap_viz = potential_new_velocities_magnitude_viz[in_bounce_gap_window_viz]

if np.sum(in_bounce_gap_window_viz) > 0:
    probability_of_failure_viz = (velocities_in_bounce_gap_viz - BOUNCE_GAP_LOWER_MPS) / \
                                 (BOUNCE_GAP_UPPER_MPS - BOUNCE_GAP_LOWER_MPS)
    random_draw_viz = np.random.rand(np.sum(in_bounce_gap_window_viz))
    probabilistic_failed_in_gap_viz = random_draw_viz < probability_of_failure_viz

    in_bounce_gap_indices_viz = np.where(in_bounce_gap_window_viz)[0]
    failed_mask_viz_temp[in_bounce_gap_indices_viz[probabilistic_failed_in_gap_viz]] = True
    successful_mask_viz_temp[in_bounce_gap_indices_viz[~probabilistic_failed_in_gap_viz]] = True

successful_mask_viz = successful_mask_viz_temp
failed_mask_viz = failed_mask_viz_temp

# Create the 3D plot
fig = plt.figure(figsize=(18, 8))

# Subplot 1: Initial Positions
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(
    base_particles_for_viz['position'][:, 0],
    base_particles_for_viz['position'][:, 1],
    base_particles_for_viz['position'][:, 2],
    color='grey', label='Initial Positions', alpha=0.5, s=20
)
ax1.set_title('Initial Particle Positions')
ax1.set_xlabel('X Position')
ax1.set_ylabel('Y Position')
ax1.set_zlabel('Z Position')
ax1.legend()

# Subplot 2: Final Positions (with color coding for success/failure) and Velocity Vectors
ax2 = fig.add_subplot(122, projection='3d')

# NEW: Create masks for successful/failed and sheep-dog/non-sheepdog particles
is_sheepdog_viz = base_particles_for_viz['is_sheepdog'] # Get the sheepdog status from the initial particles

successful_non_sheepdog_mask = successful_mask_viz & ~is_sheepdog_viz
failed_non_sheepdog_mask = failed_mask_viz & ~is_sheepdog_viz
successful_sheepdog_mask = successful_mask_viz & is_sheepdog_viz
failed_sheepdog_mask = failed_mask_viz & is_sheepdog_viz


# Plot successful non-sheepdog particles
if np.sum(successful_non_sheepdog_mask) > 0:
    ax2.scatter(
        base_particles_for_viz['position'][successful_non_sheepdog_mask, 0],
        base_particles_for_viz['position'][successful_non_sheepdog_mask, 1],
        base_particles_for_viz['position'][successful_non_sheepdog_mask, 2],
        color='green', label='Successful (Non-Sheepdog)', alpha=0.6, s=20
    )

# Plot failed non-sheepdog particles
if np.sum(failed_non_sheepdog_mask) > 0:
    ax2.scatter(
        base_particles_for_viz['position'][failed_non_sheepdog_mask, 0],
        base_particles_for_viz['position'][failed_non_sheepdog_mask, 1],
        base_particles_for_viz['position'][failed_non_sheepdog_mask, 2],
        color='red', label='Failed (Non-Sheepdog)', alpha=0.6, s=20
    )

# Plot successful sheep-dog particles
if np.sum(successful_sheepdog_mask) > 0:
    ax2.scatter(
        base_particles_for_viz['position'][successful_sheepdog_mask, 0],
        base_particles_for_viz['position'][successful_sheepdog_mask, 1],
        base_particles_for_viz['position'][successful_sheepdog_mask, 2],
        color='cyan', label='Successful (Sheep-Dog)', alpha=0.8, s=30, marker='D' # Distinct marker for sheep-dogs
    )

# Plot failed sheep-dog particles
if np.sum(failed_sheepdog_mask) > 0:
    ax2.scatter(
        base_particles_for_viz['position'][failed_sheepdog_mask, 0],
        base_particles_for_viz['position'][failed_sheepdog_mask, 1],
        base_particles_for_viz['position'][failed_sheepdog_mask, 2],
        color='magenta', label='Failed (Sheep-Dog)', alpha=0.8, s=30, marker='X' # Distinct marker for sheep-dogs
    )

# Add velocity vectors for ALL successful particles (both sheep-dog and non-sheepdog)
all_successful_mask = successful_mask_viz # This is already the combined mask
if np.sum(all_successful_mask) > 0:
    successful_positions = base_particles_for_viz['position'][all_successful_mask]
    successful_velocities_for_plot = potential_new_velocities_vector_viz[all_successful_mask] # Use the velocities calculated earlier

    # Scale vectors for better visibility
    vector_scale = 10 # Adjust as needed
    ax2.quiver(
        successful_positions[:, 0], successful_positions[:, 1], successful_positions[:, 2],
        successful_velocities_for_plot[:, 0] * vector_scale, successful_velocities_for_plot[:, 1] * vector_scale, successful_velocities_for_plot[:, 2] * vector_scale,
        color='blue', alpha=0.8, length=1, normalize=True, arrow_length_ratio=0.5
    )


# Update Title and Legend
ax2.set_title(f'Particle State at Laminar Maximum Release (Magnitude: {laminar_max_release_magnitude:.2f})\nwith Sheep-Dog Particles and Velocity Vectors (Soft Bounce-Gap)')
ax2.set_xlabel('X Position')
ax2.set_ylabel('Y Position')
ax2.set_zlabel('Z Position')
ax2.legend()

plt.tight_layout()
plt.show()

print(f"\n3D Visualization complete for Vortex Release Power: {laminar_max_release_magnitude:.2f}")
print(f"  Successful Particles: {successful_count_viz}")
print(f"  Failed Particles (converted to heat): {failed_count_viz}")
print(f"  Effective Kinetic Energy: {effective_ke_viz:.2e} Joules (from Vortex Wind-up Output)")
print(f"  Residue Heat Energy: {residue_he_viz:.2e} Joules")
```

```python
# @title
import numpy as np
import matplotlib.pyplot as plt
import copy

# Set a dark background style for plots
plt.style.use('dark_background')

# --- Simulation function integrating all forces for the sweep ---
def simulate_dynamics_for_sheepdog_sweep(
    initial_particles,
    zero_manifold_anchor,
    coherence_level,
    pull_strength_magnitude,
    stabilizer_fraction,
    correction_strength,
    time_steps,
    dt,
    max_particle_velocity=100.0
):
    particles = copy.deepcopy(initial_particles)

    # Assign sheepdog status based on stabilizer_fraction
    num_sheepdogs = int(len(particles['position']) * stabilizer_fraction)
    sheepdog_indices = np.random.choice(len(particles['position']), num_sheepdogs, replace=False)
    particles['is_sheepdog'] = np.full(len(particles['position']), False, dtype=bool)
    particles['is_sheepdog'][sheepdog_indices] = True

    # Initialize history for metrics
    lattice_tension_history = []
    coherence_history = []

    for step in range(time_steps):
        # 1. Apply Zero-Manifold Pull (if applicable)
        # Note: For the sheepdog test, we might want to keep particles 'free' or apply a background pull
        # For simplicity, let's assume a constant background pull exists or particles are just moving naturally
        # and sheepdog's role is to manage their spread/alignment.
        # If we were to apply a pull, it would be:
        # particles, _ = apply_zero_manifold_pull(particles, zero_manifold_anchor, coherence_level, pull_strength_magnitude, dt, PARTICLE_MASS_PER_PARTICLE_KG)

        # 2. Apply Vector Contagion (placeholder)
        particles = apply_vector_contagion(particles, contagion_factor=0.1) # Placeholder contagion

        # 3. Apply Sheepdog Correction
        particles = apply_sheepdog_correction(particles, correction_strength=correction_strength,
                                              boundary_min=(0,0,0), boundary_max=(100,100,100))

        # Update positions based on velocities
        particles['position'] += particles['velocity'] * dt

        # Clamp velocities to max_particle_velocity to prevent instability
        current_velocities_magnitude = np.linalg.norm(particles['velocity'], axis=1)
        clamped_velocities_mask = current_velocities_magnitude > max_particle_velocity
        if np.any(clamped_velocities_mask):
            particles['velocity'][clamped_velocities_mask] = \
                particles['velocity'][clamped_velocities_mask] / current_velocities_magnitude[clamped_velocities_mask, np.newaxis] \
                * max_particle_velocity

        # Recalculate metrics
        lattice_tension_history.append(calculate_lattice_tension(particles))
        coherence_history.append(calculate_coherence(particles))

    return particles, np.mean(lattice_tension_history), np.mean(coherence_history)


# --- Define sweep parameters ---
stabilizer_fractions = np.linspace(0.0, 0.2, 21) # From 0% to 20% in 21 steps
sweep_time_steps = 100 # Steps for each simulation run
sweep_dt = 0.1
correction_strength_sweep = 0.005 # Reduced from 0.05 for stability
pull_strength_magnitude_for_sheepdog_sim = 0.0 # No strong tractor beam for this test
coherence_level_for_sheepdog_sim = 1.0 # Assume high coherence for background

# Redefine zero_manifold_anchor and its dependencies here to ensure it's in scope
baseline_environment_temp = 300.0  # Kelvin
zero_manifold_delta_temp = 250.0  # Kelvin
zero_manifold_position = np.array([50, 50, 50]) # Centrally located in a 100x100x100 grid

zero_manifold_anchor = define_zero_manifold_anchor(
    baseline_environment_temp,
    zero_manifold_delta_temp,
    zero_manifold_position
)

# --- Initialize base particles for the sweep ---
base_particles_for_sheepdog_sweep = {}
base_particles_for_sheepdog_sweep['position'] = np.random.rand(num_particles, 3) * 100 # Random positions in 100x100x100 grid
base_particles_for_sheepdog_sweep['velocity'] = np.random.uniform(-1, 1, size=(num_particles, 3)) # Small random initial velocities
# 'is_sheepdog' will be assigned inside the simulation function for each fraction

# --- Store results ---
average_lattice_tensions = []
average_coherences = []

# --- Run the sweep ---
print(f"Starting sheepdog stabilizer fraction sweep for {len(stabilizer_fractions)} values...")
for i, frac in enumerate(stabilizer_fractions):
    final_particles_state, avg_tension, avg_coherence = simulate_dynamics_for_sheepdog_sweep(
        base_particles_for_sheepdog_sweep,
        zero_manifold_anchor, # Using the previously defined anchor, though its pull is minimal here
        coherence_level_for_sheepdog_sim,
        pull_strength_magnitude_for_sheepdog_sim,
        frac,
        correction_strength_sweep,
        sweep_time_steps,
        sweep_dt
    )
    average_lattice_tensions.append(avg_tension)
    average_coherences.append(avg_coherence)

    if (i + 1) % 5 == 0:
        print(f"  Processed {i+1}/{len(stabilizer_fractions)} fractions (Stabilizer: {frac:.2f})")

print("Sheepdog sweep complete.")

# --- Plotting the results ---
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot 1: Average Lattice Tension vs. Stabilizer Fraction
ax1.plot(stabilizer_fractions, average_lattice_tensions, marker='o', color='purple', label='Average Lattice Tension')
ax1.set_title('Average Lattice Tension vs. Stabilizer Fraction')
ax1.set_xlabel('Stabilizer Fraction')
ax1.set_ylabel('Average Lattice Tension')
ax1.grid(True)
ax1.legend()

# Plot 2: Average Coherence vs. Stabilizer Fraction
ax2.plot(stabilizer_fractions, average_coherences, marker='x', color='teal', label='Average Coherence')
ax2.set_title('Average Coherence vs. Stabilizer Fraction')
ax2.set_xlabel('Stabilizer Fraction')
ax2.set_ylabel('Average Coherence (Cosine Similarity)')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```
