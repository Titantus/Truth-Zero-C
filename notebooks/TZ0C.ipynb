````
# @title Operational Dashboard
# Cell ID: f876ba31
# Purpose: Interactive dashboard for TZ0C_Physics_Engine. Edit this cell to modify UI elements or plot logic.
# Instructions: For future modifications, reference this cell by its ID (f876ba31).
# This cell contains the interactive widgets and the dashboard plotting function.

# Set a dark background style for plots
plt.style.use('dark_background')

# Assuming TZ0C_Physics_Engine and 'engine' instance are defined in a prior cell

mode_select = widgets.Dropdown(
    options=['Ignition', 'Tuning', 'Siphon Lock'],
    value='Ignition',
    description='Engine Mode:',
)

initial_angle_widget = widgets.FloatSlider(
    value=90.0,
    min=0.0,
    max=180.0,
    step=1.0,
    description='Initial Angle:',
    continuous_update=False
)

noise_level_widget = widgets.FloatSlider(
    value=0.01,
    min=0.0,
    max=0.1,
    step=0.001,
    description='Noise Level:',
    continuous_update=False
)

bari_height_widget = widgets.FloatSlider(
    value=0.1,
    min=0.0,
    max=1.0,
    step=0.01,
    description='Barrier Height:',
    continuous_update=False
)

time_step_widget = widgets.FloatSlider(
    value=0.1,
    min=0.01,
    max=1.0,
    step=0.01,
    description='Time Step:',
    continuous_update=False
)

# New widgets for Z-Value and BOUNCE_GAP_precision
z_value_widget = widgets.IntSlider(
    value=29, # Copper
    min=1,
    max=92, # Up to Uranium (natural elements)
    step=1,
    description='Z Value (Atomic No.):',
    continuous_update=False
)

bounce_gap_precision_widget = widgets.FloatSlider(
    value=engine.BOUNCE_GAP,
    min=0.05,
    max=0.25,
    step=0.001,
    description='Bounce Gap Precision:',
    continuous_update=False
)

# New widget for pulse frequency (for 'Single Sine' pattern)
pulse_frequency_widget = widgets.FloatSlider(
    value=60.0,
    min=1.0,
    max=200.0,
    step=1.0,
    description='Pulse Frequency (Hz):',
    continuous_update=False
)

# --- New widgets for Bipolar Triple-Tap Pattern ---
pattern_type_select = widgets.Dropdown(
    options=['single_sine', 'bipolar_triple_tap'],
    value='single_sine',
    description='Input Pattern Type:',
    continuous_update=False
)

forward_tap_freqs_widget = widgets.Text(
    value='50, 75, 100',
    description='Forward Tap Freqs (Hz):',
    continuous_update=False
)

reverse_tap_freqs_widget = widgets.Text(
    value='50, 75, 100',
    description='Reverse Tap Freqs (Hz):',
    continuous_update=False
)

pivot_duration_widget = widgets.FloatSlider(
    value=0,
    min=0,
    max=100000,
    step=100,
    description='Pivot Duration (µs):',
    continuous_update=False
)

# Function to parse comma-separated frequency strings into lists of floats
def parse_freq_string(freq_str):
    try:
        return [float(f.strip()) for f in freq_str.split(',') if f.strip()]
    except ValueError:
        return [50.0, 75.0, 100.0] # Default if parsing fails

def run_dashboard(mode, initial_angle, noise_level, barrier_height, time_step, z_value,
                  bounce_gap_precision, pulse_frequency, pattern_type,
                  forward_tap_freqs_str, reverse_tap_freqs_str, pivot_duration_us):

    print(f"Executing {mode} Sequence...")

    # Set BOUNCE_GAP for the engine instance
    engine.BOUNCE_GAP = bounce_gap_precision

    # Parse frequency strings for bipolar triple-tap
    forward_tap_freqs = parse_freq_string(forward_tap_freqs_str)
    reverse_tap_freqs = parse_freq_string(reverse_tap_freqs_str)

    # Update simulate call to include z_value and pulse_frequency, and unpack new return values
    data, v_siphons_raw, led_currents_raw = engine.simulate(
        steps=2000,
        initial_angle=initial_angle,
        noise_level=noise_level,
        barrier_height_param=barrier_height,
        time_step=time_step,
        z_value=z_value,
        pulse_frequency=pulse_frequency, # Only used if pattern_type is 'single_sine'
        pattern_type=pattern_type,
        forward_tap_freqs=forward_tap_freqs,
        reverse_tap_freqs=reverse_tap_freqs,
        pivot_duration_us=pivot_duration_us
    )

    # Convert raw siphon/LED data to pandas Series or DataFrame column for consistent plotting
    data['Siphon_Potential'] = v_siphons_raw
    data['LED_Current'] = led_currents_raw

    # Create a figure with two subplots
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

    # Plot 1: Geometric Angle (ax1)
    ax1.plot(data['Step'], data['Angle'], color='cyan', label='Process Angle')
    ax1.axhline(engine.theta_lock, color='red', linestyle='--', label='Tetra-Lock')
    ax1.axhline(engine.theta_lock + 1, color='gold', linestyle=':', label='Potential Minima (+1)')
    ax1.axhline(engine.theta_lock - 1, color='gold', linestyle=':', label='Potential Minima (-1)')

    ax1.set_ylabel("Angle (Degrees)")
    ax1.legend(loc='upper right')
    ax1.set_title(f"Z={z_value} | Bounce Gap={bounce_gap_precision:.4f} | {mode} Mode | Pattern: {pattern_type} - Process Angle")
    ax1.grid(True, alpha=0.3)

    # Plot 2: Siphon Potential and LED Current (ax2)
    ax2.plot(data['Step'], data['Siphon_Potential'], color='blue', label='Siphon Potential (V)')
    ax2.plot(data['Step'], data['LED_Current'], color='lime', label='LED Current (mA)')
    ax2.axhline(y=engine.led_vf, color='red', linestyle='--', label=f'LED V_f ({engine.led_vf}V)')

    ax2.set_xlabel("Steps (Resolution)")
    ax2.set_ylabel("Voltage / Current")
    ax2.legend(loc='upper right')
    ax2.set_title(f"Z={z_value} | Pattern: {pattern_type} | {'Pulse Freq='+str(pulse_frequency)+'Hz' if pattern_type=='single_sine' else 'Pivot Dur='+str(pivot_duration_us)+'us'} | {mode} Mode - Siphon & LED Output")
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

# Group widgets for conditional visibility
triple_tap_widgets = [forward_tap_freqs_widget, reverse_tap_freqs_widget, pivot_duration_widget]

def update_pattern_widgets_visibility(pattern_value):
    if pattern_value == 'bipolar_triple_tap':
        pulse_frequency_widget.layout.display = 'none'
        for widget in triple_tap_widgets:
            widget.layout.display = 'block'
    else: # 'single_sine'
        pulse_frequency_widget.layout.display = 'block'
        for widget in triple_tap_widgets:
            widget.layout.display = 'none'

# Observe changes in pattern_type_select
pattern_type_select.observe(lambda change: update_pattern_widgets_visibility(change.new), names='value')

# Set initial visibility (run once at startup)
update_pattern_widgets_visibility(pattern_type_select.value)


# Display all controls in a VBox
controls = widgets.VBox([
    mode_select,
    initial_angle_widget,
    noise_level_widget,
    bari_height_widget,
    time_step_widget,
    z_value_widget,
    bounce_gap_precision_widget,
    pattern_type_select,
    pulse_frequency_widget,
    forward_tap_freqs_widget,
    reverse_tap_freqs_widget,
    pivot_duration_widget
])

# Use interact_manual to have a button for execution
# The layout.display property of widgets is modified directly through the observer function
# The interact call will still show all widgets, but their visibility is controlled by the observer.
interact(
    run_dashboard,
    mode=mode_select,
    initial_angle=initial_angle_widget,
    noise_level=noise_level_widget,
    barrier_height=bari_height_widget,
    time_step=time_step_widget,
    z_value=z_value_widget,
    bounce_gap_precision=bounce_gap_precision_widget,
    pulse_frequency=pulse_frequency_widget,
    pattern_type=pattern_type_select,
    forward_tap_freqs_str=forward_tap_freqs_widget,
    reverse_tap_freqs_str=reverse_tap_freqs_widget,
    pivot_duration_us=pivot_duration_widget
);
````
