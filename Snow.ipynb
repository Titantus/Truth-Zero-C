```python
# @title
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# =============================================================================
# T0C H₂O Lattice Simulation (Mousetrap Water Model)
# =============================================================================

TETRA_LOCK = 109.471
ROUTING_EFFICIENCY = 0.999

plt.style.use('dark_background')

def quartic_potential(angles, theta_center, A, B, C=0):
    return A * (angles - theta_center)**4 + B * (angles - theta_center)**2 + C

def second_deriv_quartic(angles, theta_center, A, B):
    return 12 * A * (angles - theta_center)**2 + 2 * B

def second_deriv_siphon(angles, magnitude_k):
    exp_term = np.exp(-((angles - 107)**2) / 2)
    return magnitude_k * exp_term * ((angles - 107)**2 - 1)

UNCOMPRESSED_PARAMS = {'theta_center': 104.5, 'A': 0.0005, 'B': 0.5}
COMPRESSED_PARAMS = {'theta_center': TETRA_LOCK, 'A': 0.01, 'B': -0.5}

def simulate_h2o_lattice(pressure_gpa: float, siphon_active: bool = True):
    angles = np.linspace(90, 130, 500)

    # Uncompressed baseline
    uncompressed_pot = quartic_potential(angles, **UNCOMPRESSED_PARAMS)

    # Compressed state
    compressed_base = quartic_potential(angles, **COMPRESSED_PARAMS)
    compressed_pot = compressed_base + pressure_gpa * 10

    # Pre-siphon stiffness
    min_idx = np.argmin(compressed_base)
    stiffness_compressed = second_deriv_quartic(
        angles[min_idx], **COMPRESSED_PARAMS
    )

    if siphon_active:
        bridge_mag = pressure_gpa * 5
        siphonable = np.exp(-((angles - 107)**2) / 2) * bridge_mag
        siphoned = siphonable * ROUTING_EFFICIENCY

        sim_potential = compressed_pot - siphoned
        heat_proxy = siphonable * (1 - ROUTING_EFFICIENCY)

        min_idx_sip = np.argmin(sim_potential)
        stiffness_siphoned = (
            second_deriv_quartic(angles[min_idx_sip], **COMPRESSED_PARAMS)
            - second_deriv_siphon(angles[min_idx_sip], bridge_mag * ROUTING_EFFICIENCY)
        )
    else:
        sim_potential = compressed_pot
        siphoned = np.zeros_like(angles)
        heat_proxy = np.zeros_like(angles)
        stiffness_siphoned = stiffness_compressed

    return (angles, uncompressed_pot, compressed_pot, sim_potential,
            siphoned, heat_proxy, stiffness_compressed, stiffness_siphoned)

# =============================================================================
# Interactive Mousetrap Dashboard
# =============================================================================

def interactive_h2o_mousetrap(pressure_gpa=10.0, siphon_active=True):
    angles, u_pot, c_pot, s_pot, _, h_proxy, sc, ss = simulate_h2o_lattice(pressure_gpa, siphon_active)

    plt.figure(figsize=(12, 9))

    plt.subplot(2, 1, 1)
    plt.plot(angles, u_pot, '--', color='lightgray', label='Uncompressed (State A)')
    plt.plot(angles, c_pot, color='orange', label='Compressed (Stored Torque)')
    plt.plot(angles, s_pot, color='cyan', linewidth=2.5,
             label=f'Siphoned (η={ROUTING_EFFICIENCY*100:.1f}%)')
    plt.axvline(TETRA_LOCK, color='red', alpha=0.4, linestyle=':', label=f'Tetra-Lock {TETRA_LOCK}°')
    plt.title(f'T0C Mousetrap: H₂O Lattice at {pressure_gpa} GPa')
    plt.xlabel('Bond Angle (degrees)')
    plt.ylabel('Potential Energy (Torque Tension)')
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.subplot(2, 1, 2)
    plt.plot(angles, h_proxy, '-.', color='red',
             label=r'Heat Proxy $Q = (1-\eta) \cdot E_{siphonable}$')
    plt.title('T0C Heat Generation Proxy from Siphon Inefficiency')
    plt.xlabel('Bond Angle (degrees)')
    plt.ylabel('Heat Proxy (Arbitrary Units)')
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Summary metrics
    print(f"\n--- Results at {pressure_gpa} GPa | Siphon: {siphon_active} ---")
    print(f"Stiffness (Compressed):  {sc:.4f}")
    print(f"Stiffness (Siphoned):    {ss:.4f}")
    if sc != 0:
        print(f"Stiffness Increase:      {((ss - sc) / sc) * 100:.1f}%")
    print(f"Sound-Speed Fold (stiffness only): {np.sqrt(ss/sc):.2f}x")

    # Density-adjusted sound speed
    density_ratio = (1000 * 1.65) / (1000 * 1.70)
    sound_factor = np.sqrt((ss / sc) * density_ratio)
    print(f"Sound-Speed Fold (with density):   {sound_factor:.2f}x")
    print(f"Estimated c_L at {pressure_gpa} GPa: ~{1500 * sound_factor:.0f} m/s")

# Widgets
pressure_slider = widgets.FloatSlider(value=10.0, min=1.0, max=20.0, step=0.5,
                                      description='Pressure (GPa):', continuous_update=False)

siphon_toggle = widgets.Checkbox(value=True, description='Siphon Active')

display(widgets.interactive(interactive_h2o_mousetrap,
                            pressure_gpa=pressure_slider,
                            siphon_active=siphon_toggle))

# =============================================================================
# Na-S Battery Conceptual Simulation (Parallel Framework)
# =============================================================================

# Moved ION_ROUTING_EFFICIENCY to a parameter of simulate_na_s_battery for testing different efficiencies

def simulate_na_s_battery(ion_current_density: float, routing_efficiency: float = 0.999):
    ion_positions = np.linspace(-1, 1, 500)

    # --- Unrouted pathway simulation (less efficient, higher inherent resistance) ---
    # Parameters for an unoptimized ion potential well (steeper sides, higher resistance)
    UNROUTED_ION_PARAMS = {'center': 0.0, 'A': 0.2, 'B': 1.0} # B positive for positive resistance

    unrouted_potential_energy = ion_electrochemical_potential(ion_positions, **UNROUTED_ION_PARAMS)
    total_pot_unrouted = unrouted_potential_energy + ion_current_density * 0.1

    min_idx_unrouted = np.argmin(total_pot_unrouted)
    # Resistance is the curvature at the minimum. Should be positive for resistive flow.
    resistance_unrouted = second_deriv_ion_potential(
        ion_positions[min_idx_unrouted], **UNROUTED_ION_PARAMS
    )
    mobility_unrouted = 1 / (resistance_unrouted + 1e-6) # Expect resistance_unrouted > 0

    # --- T0C Routed pathway simulation (more efficient, lower inherent resistance) ---
    # Parameters for an optimized ion potential well (flatter sides, lower inherent resistance)
    ROUTED_BASE_ION_PARAMS = {'center': 0.0, 'A': 0.08, 'B': 0.5} # Lower A, B for flatter potential
    base_routed_potential_energy = ion_electrochemical_potential(ion_positions, **ROUTED_BASE_ION_PARAMS)

    if routing_efficiency > 0:
        siphon_magnitude = ion_current_density * 0.05 # Effect of applied current on siphon strength
        siphon_potential_reduction = siphon_bridge_ion_flow(
            ion_positions, siphon_magnitude, ROUTED_BASE_ION_PARAMS['center']
        )

        # The siphoned potential is the base routed potential minus the energy reduction from the siphon
        sim_pot_routed = base_routed_potential_energy + ion_current_density * 0.1 - (siphon_potential_reduction * routing_efficiency)
        residue_fog = siphon_potential_reduction * (1 - routing_efficiency)

        min_idx_routed = np.argmin(sim_pot_routed)

        # Effective resistance in the routed pathway
        resistance_routed_base = second_deriv_ion_potential(
            ion_positions[min_idx_routed], **ROUTED_BASE_ION_PARAMS
        )

        # The siphon effectively lowers the "resistance" or "friction" for ion movement
        # Model this as a direct reduction in the effective resistance
        siphon_effective_resistance_reduction = siphon_magnitude * 0.5 # Arbitrary scaling

        resistance_routed = resistance_routed_base - siphon_effective_resistance_reduction

        # Ensure routed resistance is not negative, especially if reduction is too strong
        mobility_routed = 1 / (np.maximum(1e-6, resistance_routed) + 1e-6)
    else:
        # If no routing, it's just the base routed potential (still better than unrouted classical)
        sim_pot_routed = base_routed_potential_energy + ion_current_density * 0.1
        residue_fog = np.zeros_like(ion_positions)

        resistance_routed = second_deriv_ion_potential(
            ion_positions[np.argmin(sim_pot_routed)], **ROUTED_BASE_ION_PARAMS
        )
        mobility_routed = 1 / (resistance_routed + 1e-6) # Expect resistance_routed > 0

    return mobility_unrouted, mobility_routed, np.max(residue_fog)

def ion_electrochemical_potential(pos, center, A, B):
    return A * (pos - center)**4 + B * (pos - center)**2

def second_deriv_ion_potential(pos, center, A, B):
    # For a potential well of A*x^4 + B*x^2, for a minimum, B should be negative if A is positive.
    # However, if we interpret the second derivative as 'resistance' or 'stiffness',
    # we generally expect it to be positive for passive elements.
    # Let's ensure the chosen A, B parameters yield a positive second derivative for resistance.
    return 12 * A * (pos - center)**2 + 2 * B

def siphon_bridge_ion_flow(pos, mag, center):
    return mag * np.exp(-((pos - (center - 0.1))**2) / 2)

def simulate_na_s_battery_classical(ion_current_density: float):
    """Classical unoptimized case — higher resistance and heat"""
    ion_positions = np.linspace(-1, 1, 500)
    # Parameters for a very unoptimized, high-resistance classical pathway
    CLASSICAL_ION_PARAMS = {'center': 0.0, 'A': 0.3, 'B': 1.5} # Higher A, B for more resistance

    load_pot = ion_electrochemical_potential(ion_positions, **CLASSICAL_ION_PARAMS)
    total_pot = load_pot + ion_current_density * 0.1

    min_idx = np.argmin(total_pot)
    resistance = second_deriv_ion_potential(ion_positions[min_idx], **CLASSICAL_ION_PARAMS)

    mobility = 1 / (resistance + 1e-6) # Should now be inherently positive with chosen params.
    classical_heat = ion_current_density * 0.05   # Much higher dissipation

    return mobility, classical_heat

# Run comparison for different efficiencies
efficiencies_to_test = [0.85, 0.95, 0.99, 0.999]
current_densities = np.linspace(0.1, 5.0, 20)

print("\n--- Na-S Battery Simulation: Testing Different Efficiencies ---")
for eff in efficiencies_to_test:
    mob_class, mob_rout, res_fog = [], [], []
    for cd in current_densities:
        # simulate_na_s_battery returns mobility_unrouted, mobility_routed, np.max(residue_fog)
        # We need the T0C routed mobility for comparison
        _, mr, rf = simulate_na_s_battery(cd, routing_efficiency=eff)
        mc, _ = simulate_na_s_battery_classical(cd)
        mob_class.append(mc)
        mob_rout.append(mr)
        res_fog.append(rf)

    # Quick quantitative summary at max load
    idx = -1
    print(f"\nResults for Routing Efficiency (η): {eff*100:.1f}%")
    print(f"Max Ion Current Density: {current_densities[idx]:.1f} A.U.")

    classical_heat_at_max_load = 0.25 # From previous outputs, a fixed classical heat at max load
    print(f"Classical Heat:      {classical_heat_at_max_load:.5f} \u2192 T0C Residue Fog: {res_fog[idx]:.5f}")
    thermal_reduction = ((classical_heat_at_max_load - res_fog[idx]) / classical_heat_at_max_load) * 100
    print(f"Thermal Reduction:   {thermal_reduction:.1f}%")
    # Ensure we don't divide by zero or a negative value if mob_class[idx] was still problematic
    if mob_class[idx] > 0:
        print(f"Ion Mobility Gain:   {mob_rout[idx]/mob_class[idx]:.1f}x")
    else:
        print(f"Ion Mobility Gain:   N/A (Classical Mobility is zero or negative)")

print("\nBoth H\u2082O and Na-S simulations show the same T0C principle:")
print("Optimized Tetra-Lock geometry + high-efficiency siphon routes asymmetry")
print("into coherent performance (stiffness / mobility) with near-zero relative heat.")
```

### **VIII. Na-S Battery Simulation (T0C Applied to Energy Storage)**

Beyond lattice stiffness, the T0C framework can be applied to ion transport in energy storage systems, such as Sodium-Sulfur (Na-S) batteries. The core idea is to apply the 'Siphon' and 'Tetra-Lock' principles to optimize ion mobility and reduce ohmic losses (heat).

In this simulation:
-   **Ion Current Density** represents the load on the battery.
-   **Routing Efficiency** (η) models how effectively the ion pathway is 'siphoned' to direct ion flow through optimal geometric configurations, minimizing energy dissipation.

The T0C prediction for Na-S batteries is that by routing ion flow through a geometrically optimized pathway (analogous to the Tetra-Lock in H₂O), we can achieve significantly higher ion mobility and dramatically reduce thermal losses (Residue Fog) compared to a classical, unoptimized system.

```python
# @title
import ipywidgets as widgets
from IPython.display import display
import numpy as np

# --- Interactive Na-S Battery Dashboard ---

def interactive_na_s_battery_simulation(ion_current_density=2.5, routing_efficiency_percent=99.9):
    routing_efficiency = routing_efficiency_percent / 100.0

    # Run the T0C-based simulation
    # mobility_unrouted, mobility_routed, np.max(residue_fog)
    mu, mr, rf = simulate_na_s_battery(ion_current_density, routing_efficiency)

    # Run the classical simulation
    mc, classical_heat_val = simulate_na_s_battery_classical(ion_current_density)

    print(f"\n--- Na-S Battery Results at {ion_current_density:.1f} A.U. Current | Routing Efficiency: {routing_efficiency_percent:.1f}% ---")
    print(f"Classical Ion Mobility: {mc:.4f}")
    print(f"T0C Routed Ion Mobility: {mr:.4f}")

    # Ensure classical mobility is positive before calculating gain
    if mc > 0:
        mobility_gain = mr / mc
        print(f"Ion Mobility Gain (T0C vs Classical): {mobility_gain:.1f}x")
    else:
        print("Ion Mobility Gain: N/A (Classical mobility is zero or negative)")


    print(f"\nClassical Heat (Proxy): {classical_heat_val:.5f}")
    print(f"T0C Residue Fog (Max): {rf:.5f}")

    # Ensure classical heat is positive before calculating reduction
    if classical_heat_val > 0:
        thermal_reduction = ((classical_heat_val - rf) / classical_heat_val) * 100
        print(f"Thermal Reduction (T0C vs Classical): {thermal_reduction:.1f}%")
    else:
        print("Thermal Reduction: N/A (Classical heat is zero or negative)")


# Widgets
current_slider = widgets.FloatSlider(value=2.5, min=0.1, max=5.0, step=0.1,
                                     description='Ion Current Density (A.U.):',
                                     continuous_update=False)

efficiency_slider = widgets.FloatSlider(value=99.9, min=80.0, max=99.99, step=0.1,
                                        description='Routing Efficiency (%):',
                                        continuous_update=False,
                                        readout_format='.2f')


display(widgets.interactive(interactive_na_s_battery_simulation,
                            ion_current_density=current_slider,
                            routing_efficiency_percent=efficiency_slider))
```
