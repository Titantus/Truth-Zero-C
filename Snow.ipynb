```python
# @title
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# =============================================================================
# T0C H₂O Lattice Simulation (Mousetrap Water Model)
# =============================================================================

TETRA_LOCK = 109.471
ROUTING_EFFICIENCY = 0.999

plt.style.use('dark_background')

def quartic_potential(angles, theta_center, A, B, C=0):
    return A * (angles - theta_center)**4 + B * (angles - theta_center)**2 + C

def second_deriv_quartic(angles, theta_center, A, B):
    return 12 * A * (angles - theta_center)**2 + 2 * B

def second_deriv_siphon(angles, magnitude_k):
    exp_term = np.exp(-((angles - 107)**2) / 2)
    return magnitude_k * exp_term * ((angles - 107)**2 - 1)

UNCOMPRESSED_PARAMS = {'theta_center': 104.5, 'A': 0.0005, 'B': 0.5}
COMPRESSED_PARAMS = {'theta_center': TETRA_LOCK, 'A': 0.01, 'B': -0.5}

def simulate_h2o_lattice(pressure_gpa: float, siphon_active: bool = True):
    angles = np.linspace(90, 130, 500)

    # Uncompressed baseline
    uncompressed_pot = quartic_potential(angles, **UNCOMPRESSED_PARAMS)

    # Compressed state
    compressed_base = quartic_potential(angles, **COMPRESSED_PARAMS)
    compressed_pot = compressed_base + pressure_gpa * 10

    # Pre-siphon stiffness
    min_idx = np.argmin(compressed_base)
    stiffness_compressed = second_deriv_quartic(
        angles[min_idx], **COMPRESSED_PARAMS
    )

    if siphon_active:
        bridge_mag = pressure_gpa * 5
        siphonable = np.exp(-((angles - 107)**2) / 2) * bridge_mag
        siphoned = siphonable * ROUTING_EFFICIENCY

        sim_potential = compressed_pot - siphoned
        heat_proxy = siphonable * (1 - ROUTING_EFFICIENCY)

        min_idx_sip = np.argmin(sim_potential)
        stiffness_siphoned = (
            second_deriv_quartic(angles[min_idx_sip], **COMPRESSED_PARAMS)
            - second_deriv_siphon(angles[min_idx_sip], bridge_mag * ROUTING_EFFICIENCY)
        )
    else:
        sim_potential = compressed_pot
        siphoned = np.zeros_like(angles)
        heat_proxy = np.zeros_like(angles)
        stiffness_siphoned = stiffness_compressed

    return (angles, uncompressed_pot, compressed_pot, sim_potential,
            siphoned, heat_proxy, stiffness_compressed, stiffness_siphoned)

# =============================================================================
# Interactive Mousetrap Dashboard
# =============================================================================

def interactive_h2o_mousetrap(pressure_gpa=10.0, siphon_active=True):
    angles, u_pot, c_pot, s_pot, _, h_proxy, sc, ss = simulate_h2o_lattice(pressure_gpa, siphon_active)

    plt.figure(figsize=(12, 9))

    plt.subplot(2, 1, 1)
    plt.plot(angles, u_pot, '--', color='lightgray', label='Uncompressed (State A)')
    plt.plot(angles, c_pot, color='orange', label='Compressed (Stored Torque)')
    plt.plot(angles, s_pot, color='cyan', linewidth=2.5,
             label=f'Siphoned (η={ROUTING_EFFICIENCY*100:.1f}%)')
    plt.axvline(TETRA_LOCK, color='red', alpha=0.4, linestyle=':', label=f'Tetra-Lock {TETRA_LOCK}°')
    plt.title(f'T0C Mousetrap: H₂O Lattice at {pressure_gpa} GPa')
    plt.xlabel('Bond Angle (degrees)')
    plt.ylabel('Potential Energy (Torque Tension)')
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.subplot(2, 1, 2)
    plt.plot(angles, h_proxy, '-.', color='red',
             label=r'Heat Proxy $Q = (1-\eta) \cdot E_{siphonable}$')
    plt.title('T0C Heat Generation Proxy from Siphon Inefficiency')
    plt.xlabel('Bond Angle (degrees)')
    plt.ylabel('Heat Proxy (Arbitrary Units)')
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Summary metrics
    print(f"\n--- Results at {pressure_gpa} GPa | Siphon: {siphon_active} ---")
    print(f"Stiffness (Compressed):  {sc:.4f}")
    print(f"Stiffness (Siphoned):    {ss:.4f}")
    if sc != 0:
        print(f"Stiffness Increase:      {((ss - sc) / sc) * 100:.1f}%")
    print(f"Sound-Speed Fold (stiffness only): {np.sqrt(ss/sc):.2f}x")

    # Density-adjusted sound speed
    density_ratio = (1000 * 1.65) / (1000 * 1.70)
    sound_factor = np.sqrt((ss / sc) * density_ratio)
    print(f"Sound-Speed Fold (with density):   {sound_factor:.2f}x")
    print(f"Estimated c_L at {pressure_gpa} GPa: ~{1500 * sound_factor:.0f} m/s")

# Widgets
pressure_slider = widgets.FloatSlider(value=10.0, min=1.0, max=20.0, step=0.5,
                                      description='Pressure (GPa):', continuous_update=False)

siphon_toggle = widgets.Checkbox(value=True, description='Siphon Active')

display(widgets.interactive(interactive_h2o_mousetrap,
                            pressure_gpa=pressure_slider,
                            siphon_active=siphon_toggle))

# =============================================================================
# Na-S Battery Conceptual Simulation (Parallel Framework)
# =============================================================================

# Moved ION_ROUTING_EFFICIENCY to a parameter of simulate_na_s_battery for testing different efficiencies

# Helper functions for ion potential
def ion_electrochemical_potential(pos, center, A, B):
    return A * (pos - center)**4 + B * (pos - center)**2

def second_deriv_ion_potential(pos, center, A, B):
    return 12 * A * (pos - center)**2 + 2 * B

def siphon_bridge_ion_flow(pos, mag, center):
    return mag * np.exp(-((pos - (center - 0.1))**2) / 2)

# Parameters for ion potential interpolation
UNROUTED_ION_PARAMS = {'center': 0.0, 'A': 0.2, 'B': 1.0} # Baseline unoptimized parameters
TARGET_OPTIMIZED_ION_PARAMS = {'center': 0.0, 'A': 0.08, 'B': 0.5} # Target optimized parameters (like Oloid)

def simulate_na_s_battery(
    ion_current_density: float,
    routing_efficiency: float = 0.999,
    geometry_optimization_level: float = 1.0 # New parameter for Oloid concepts
):
    ion_positions = np.linspace(-1, 1, 500)

    # --- Unrouted pathway simulation (for comparison) ---
    unrouted_potential_energy = ion_electrochemical_potential(ion_positions, **UNROUTED_ION_PARAMS)
    total_pot_unrouted = unrouted_potential_energy + ion_current_density * 0.1
    min_idx_unrouted = np.argmin(total_pot_unrouted)
    resistance_unrouted = second_deriv_ion_potential(
        ion_positions[min_idx_unrouted], **UNROUTED_ION_PARAMS
    )
    mobility_unrouted = 1 / (resistance_unrouted + 1e-6)

    # --- T0C Routed pathway simulation with geometry optimization ---
    # Interpolate 'A' and 'B' parameters based on geometry_optimization_level
    A_current = UNROUTED_ION_PARAMS['A'] * (1 - geometry_optimization_level) + TARGET_OPTIMIZED_ION_PARAMS['A'] * geometry_optimization_level
    B_current = UNROUTED_ION_PARAMS['B'] * (1 - geometry_optimization_level) + TARGET_OPTIMIZED_ION_PARAMS['B'] * geometry_optimization_level
    current_routed_params = {'center': 0.0, 'A': A_current, 'B': B_current}

    base_routed_potential_energy = ion_electrochemical_potential(ion_positions, **current_routed_params)

    if routing_efficiency > 0:
        # Siphon magnitude also scaled by geometry optimization level
        siphon_magnitude = ion_current_density * 0.05 * geometry_optimization_level
        siphon_potential_reduction = siphon_bridge_ion_flow(
            ion_positions, siphon_magnitude, current_routed_params['center']
        )

        sim_pot_routed = base_routed_potential_energy + ion_current_density * 0.1 - (siphon_potential_reduction * routing_efficiency)
        residue_fog = siphon_potential_reduction * (1 - routing_efficiency)

        min_idx_routed = np.argmin(sim_pot_routed)

        resistance_routed_base = second_deriv_ion_potential(
            ion_positions[min_idx_routed], **current_routed_params
        )

        # Siphon effective resistance reduction also scaled
        siphon_effective_resistance_reduction = siphon_magnitude * 0.5 * geometry_optimization_level
        resistance_routed = resistance_routed_base - siphon_effective_resistance_reduction

        mobility_routed = 1 / (np.maximum(1e-6, resistance_routed) + 1e-6)
    else:
        sim_pot_routed = base_routed_potential_energy + ion_current_density * 0.1
        residue_fog = np.zeros_like(ion_positions)
        resistance_routed = second_deriv_ion_potential(
            ion_positions[np.argmin(sim_pot_routed)], **current_routed_params
        )
        mobility_routed = 1 / (resistance_routed + 1e-6)

    return mobility_unrouted, mobility_routed, np.max(residue_fog)

def simulate_na_s_battery_classical(ion_current_density: float):
    """Classical unoptimized case — higher resistance and heat"""
    ion_positions = np.linspace(-1, 1, 500)
    # Parameters for a very unoptimized, high-resistance classical pathway
    CLASSICAL_ION_PARAMS = {'center': 0.0, 'A': 0.3, 'B': 1.5} # Higher A, B for more resistance

    load_pot = ion_electrochemical_potential(ion_positions, **CLASSICAL_ION_PARAMS)
    total_pot = load_pot + ion_current_density * 0.1

    min_idx = np.argmin(total_pot)
    resistance = second_deriv_ion_potential(ion_positions[min_idx], **CLASSICAL_ION_PARAMS)

    mobility = 1 / (resistance + 1e-6) # Should now be inherently positive with chosen params.
    classical_heat = ion_current_density * 0.05   # Much higher dissipation

    return mobility, classical_heat

# Run comparison for different efficiencies
efficiencies_to_test = [0.85, 0.95, 0.99, 0.999]
current_densities = np.linspace(0.1, 5.0, 20)

print("\n--- Na-S Battery Simulation: Testing Different Efficiencies ---")
for eff in efficiencies_to_test:
    mob_class, mob_rout, res_fog = [], [], []
    for cd in current_densities:
        # simulate_na_s_battery returns mobility_unrouted, mobility_routed, np.max(residue_fog)
        # We need the T0C routed mobility for comparison
        # Using a fixed geometry_optimization_level for this batch run for now
        _, mr, rf = simulate_na_s_battery(cd, routing_efficiency=eff, geometry_optimization_level=1.0)
        mc, _ = simulate_na_s_battery_classical(cd)
        mob_class.append(mc)
        mob_rout.append(mr)
        res_fog.append(rf)

    # Quick quantitative summary at max load
    idx = -1
    print(f"\nResults for Routing Efficiency (η): {eff*100:.1f}%")
    print(f"Max Ion Current Density: {current_densities[idx]:.1f} A.U.")

    classical_heat_at_max_load = 0.25 # From previous outputs, a fixed classical heat at max load
    print(f"Classical Heat:      {classical_heat_at_max_load:.5f} \u2192 T0C Residue Fog: {res_fog[idx]:.5f}")
    thermal_reduction = ((classical_heat_at_max_load - res_fog[idx]) / classical_heat_at_max_load) * 100
    print(f"Thermal Reduction:   {thermal_reduction:.1f}%")
    # Ensure we don't divide by zero or a negative value if mob_class[idx] was still problematic
    if mob_class[idx] > 0:
        print(f"Ion Mobility Gain:   {mob_rout[idx]/mob_class[idx]:.1f}x")
    else:
        print(f"Ion Mobility Gain:   N/A (Classical Mobility is zero or negative)")

print("\nBoth H\u2082O and Na-S simulations show the same T0C principle:")
print("Optimized Tetra-Lock geometry + high-efficiency siphon routes asymmetry")
print("into coherent performance (stiffness / mobility) with near-zero relative heat.")
```

```python
# @title
# =============================================================================
# T0C Oloid Simulation (Vibration-to-Motion Conversion)
# =============================================================================

# Reusing potential functions from above, adapting them for oloid rolling dynamics

def quartic_potential(angles, theta_center, A, B, C=0):
    return A * (angles - theta_center)**4 + B * (angles - theta_center)**2 + C

def second_deriv_quartic(angles, theta_center, A, B):
    return 12 * A * (angles - theta_center)**2 + 2 * B

# Define parameters for an unoptimized rolling object and an optimized oloid
# 'deviation' from ideal straight roll is our 'angle'
UNOPTIMIZED_ROLL_PARAMS = {'theta_center': 0.0, 'A': 0.1, 'B': 0.5} # A sharper, more resistive well
OPTIMIZED_ROLL_PARAMS = {'theta_center': 0.0, 'A': 0.01, 'B': -0.1} # A flatter, less resistive well (or double-well like)

# Siphon efficiency for routing vibration into motion
OLID_SIPHON_EFFICIENCY = 0.999 # High efficiency for an ideal oloid

# T0C Constant: Conflict Threshold (Ω_c ≈ 4.0) - from T0C II.4
# We'll use this to model a 'causal pivot' if shake is too high
TOC_CONFLICT_THRESHOLD = 4.0

def simulate_oloid_roll(
    shake_amplitude: float, # Renamed from vibration_amplitude to better reflect 'shake' from T0C VI
    geometry_optimization_level: float = 1.0, # Corresponds to Definition Invariant (Axiom 12) & Prism Logic (T0C IV)
    siphon_active: bool = True,
    tetra_oloid_factor: float = 0.0 # 0 for standard oloid (flat COM), 1 for full Tetra-Oloid wobble (energy cost for COM changes)
):
    # 'deviation_from_ideal_roll' represents the unresolved asymmetry (Axiom 11) or 'wobble' (Triality, Axiom 3)
    # seeking cancellation through the oloid's rolling motion.
    deviation_from_ideal_roll = np.linspace(-3, 3, 500)

    # Interpolate parameters based on geometry_optimization_level
    # This reflects the Definition Invariant (Axiom 12): collapsing Omni into a renderable state
    # via specific geometric configuration (Prism Logic, T0C IV).
    A_current = UNOPTIMIZED_ROLL_PARAMS['A'] * (1 - geometry_optimization_level) + OPTIMIZED_ROLL_PARAMS['A'] * geometry_optimization_level
    B_current = UNOPTIMIZED_ROLL_PARAMS['B'] * (1 - geometry_optimization_level) + OPTIMIZED_ROLL_PARAMS['B'] * geometry_optimization_level
    current_roll_params = {'theta_center': 0.0, 'A': A_current, 'B': B_current}

    # Introduce wobble potential for Tetra-Oloid concept
    # The 'wobble' is an energy cost associated with a non-flat COM, scaled by tetra_oloid_factor.
    # A simplified cosine function is used to represent periodic energy fluctuation due to wobble.
    wobble_magnitude = 0.05 # Arbitrary magnitude for the wobble's energy cost
    # The phase and frequency are chosen to make the wobble noticeable and conceptually align with deviation
    wobble_potential_component = (1 - np.cos(deviation_from_ideal_roll * np.pi / 5)) * wobble_magnitude * tetra_oloid_factor

    # Baseline potential for the current geometry (without siphon or shake), now including wobble's energy cost
    # This represents the inherent resistance or "asymmetry" of the geometry itself, plus energy due to COM changes.
    baseline_potential = quartic_potential(deviation_from_ideal_roll, **current_roll_params) + wobble_potential_component

    # COM height proxy: directly related to the wobble potential introduced
    # This represents the *energy cost* associated with COM vertical movement, not literal height in meters.
    com_height_proxy = wobble_potential_component

    # Effect of 'shake_amplitude' on the system potential
    # 'shake_amplitude' can be seen as an input of unresolved asymmetry (Axiom 11) or 'Energy' (Axiom 6)
    # that the system needs to process. Higher shake means more input asymmetry.
    vibration_potential = baseline_potential + shake_amplitude * 0.1

    # --- Implement Conflict Threshold (T0C II.4) ---
    # If the shake amplitude (Ω) exceeds a conflict threshold (Ω_c), it forces a 'Causal Pivot' (T0C III.3)
    # which can lead to increased resistance or reduced siphon efficiency (more Residue Fog).
    conflict_effect_multiplier = 1.0
    if shake_amplitude >= TOC_CONFLICT_THRESHOLD:
        # Simulate a Causal Pivot: increased resistance, potential for instability.
        # This could also be a drop in siphon efficiency or a shift to a 'RESIDUE' routing mode (T0C III.2)
        conflict_effect_multiplier = 1.0 + (shake_amplitude - TOC_CONFLICT_THRESHOLD) * 0.5

    # Pre-siphon effective resistance/stiffness at the ideal roll (deviation=0)
    # This represents the "unresolved process" (Asymmetry, Axiom 11) before routing.
    min_idx_base = np.argmin(baseline_potential) # Should be near 0
    resistance_pre_siphon = second_deriv_quartic(deviation_from_ideal_roll[min_idx_base], **current_roll_params) * conflict_effect_multiplier

    if siphon_active:
        # Siphon effect: reduces potential around ideal roll, facilitating motion
        # This is the 'Siphon' (T0C III.6) actively routing asymmetry.
        # The 'geometry_optimization_level' enhances the siphon's capacity, aligning with Definition Invariant.
        siphon_magnitude = shake_amplitude * 0.05 * geometry_optimization_level # Stronger siphon with more optimization
        siphon_effect = siphon_magnitude * np.exp(-((deviation_from_ideal_roll)**2) / (0.5 + 2*(1-geometry_optimization_level))) # Localized potential reduction

        # The siphoned potential reflects the "neutralization steps" (Axiom 2)
        # and the "renderable state" (Axiom 1, DOT) achieved through routing.
        # 'OLID_SIPHON_EFFICIENCY' reflects the 'ROUTING_EFFICIENCY' in the Propagation-Consolidation Invariant (T0C III.5).
        sim_potential = vibration_potential - (siphon_effect * OLID_SIPHON_EFFICIENCY)

        # 'heat_proxy' is the 'Residue Fog' (T0C III.2, III.4) – un-neutralized asymmetry
        # dissipated as heat, a manifestation of entropy (Propagation-Consolidation Invariant III.5).
        heat_proxy = siphon_effect * (1 - OLID_SIPHON_EFFICIENCY) * conflict_effect_multiplier # Conflict increases residue

        min_idx_sim = np.argmin(sim_potential)
        resistance_siphoned = (
            second_deriv_quartic(deviation_from_ideal_roll[min_idx_sim], **current_roll_params)
            - (siphon_magnitude * 0.5 * geometry_optimization_level) # Siphon directly reduces effective resistance
        ) * conflict_effect_multiplier # Conflict increases resistance
        # Ensure resistance is not negative
        resistance_siphoned = np.maximum(1e-6, resistance_siphoned)
    else:
        sim_potential = vibration_potential
        siphon_effect = np.zeros_like(deviation_from_ideal_roll)
        heat_proxy = np.zeros_like(deviation_from_ideal_roll)
        resistance_siphoned = resistance_pre_siphon # No siphon, so no change in resistance from siphon

    # Rolling efficiency is inversely proportional to effective resistance.
    # This is a measure of "Truth <=> Renderability" (Axiom 10) and "Asymmetry <=> Existence" (Axiom 11)
    # Higher efficiency means lower resistance to maintaining ideal roll under 'shake'.
    rolling_efficiency_pre_siphon = 1 / (resistance_pre_siphon + 1e-6)
    rolling_efficiency_siphoned = 1 / (resistance_siphoned + 1e-6)

    return (
        deviation_from_ideal_roll,
        baseline_potential,
        vibration_potential,
        sim_potential,
        siphon_effect,
        heat_proxy,
        rolling_efficiency_pre_siphon,
        rolling_efficiency_siphoned,
        com_height_proxy # New return value
    )
```

### **2D Physics Analogue: Vibration-to-Motion Siphon (Cam on Vibrating Plate)**

This simulation demonstrates the core principle of the 'Vibration-to-Motion Siphon' using a simplified 2D model. Imagine an asymmetric shape (a 'cam' or 'lobe') resting on a vibrating surface. The goal is to convert the chaotic, vertical vibrational energy into directed horizontal motion.

**Analogy to T0C Concepts:**

*   **Vibrating Plate:** Represents the input of unresolved asymmetry ('shake' or external energy), much like the oloid on a vibrating surface.
*   **Asymmetric Cam:** Represents the 'geometry optimization' (Definition Invariant, Prism Logic). Its shape is crucial for routing torque.
*   **Bounce-Gap / Elasticity:** The elasticity of the cam (or the interaction with the plate) creates a 'Bounce-Gap' (T0C II.2). This detuning window is essential for enabling the conversion of vertical impulses into horizontal translation.
*   **Siphon:** The cam's geometry, interacting with the vibrating plate and its elasticity, acts as a 'Siphon' (T0C III.6). It channels the vertical energy into horizontal motion, minimizing 'Residue Fog' (lost energy/heat) and maximizing efficiency.
*   **Horizontal Velocity:** This is the 'coherent performance' or 'renderable state' (Axiom 1, DOT) achieved by efficiently routing the asymmetry.

```python
# @title
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display
import ipywidgets as widgets

# --- Simulation Parameters ---
DT = 0.01  # Time step
TOTAL_TIME = 10 # Total simulation time

# Cam parameters (simple asymmetric shape for demonstration)
CAM_RADIUS = 0.5
CAM_MASS = 1.0

# Plate parameters
PLATE_HEIGHT = -1.0 # Y-coordinate of the plate

def simulate_cam_roll(
    vibration_frequency=50.0, # Hz
    vibration_amplitude=0.1,  # Vertical amplitude of plate vibration
    cam_elasticity=0.5,       # Represents the 'bounce-gap' or material properties
    asymmetry_factor=0.3      # Controls how 'lopsided' the cam is
):
    time_points = np.arange(0, TOTAL_TIME, DT)

    # Cam state variables: [x, y, vx, vy, angle, angular_velocity]
    cam_state = np.array([0.0, CAM_RADIUS, 0.0, 0.0, 0.0, 0.0]) # Start at origin, resting on plate

    history = []

    for t in time_points:
        # Plate vibration (vertical only)
        plate_y = PLATE_HEIGHT + vibration_amplitude * np.sin(2 * np.pi * vibration_frequency * t)

        # Cam's current point of contact (simplified for a cam-like shape)
        # This simulates a cam's varying 'radius' from its center based on its angle
        effective_radius = CAM_RADIUS + asymmetry_factor * np.sin(cam_state[4]) # angle is cam_state[4]
        contact_point_y = cam_state[1] - effective_radius * np.cos(cam_state[4]) # Simplified vertical contact
        contact_point_x = cam_state[0] + effective_radius * np.sin(cam_state[4]) # Simplified horizontal contact

        # Gravity
        cam_state[3] -= 9.81 * DT

        # Collision detection and response with the vibrating plate
        if contact_point_y <= plate_y:
            # Correct position to be on the plate
            cam_state[1] += (plate_y - contact_point_y) # Adjust y to surface

            # Apply impulse for bounce (vertical component)
            if cam_state[3] < 0: # Only if moving downwards
                # Elastic collision: velocity reversal + energy loss (1-elasticity)
                cam_state[3] = -cam_state[3] * cam_elasticity

            # Horizontal impulse for rolling motion (the siphon effect)
            # This is the core 'siphon' mechanism: vertical impact routes to horizontal push
            # The asymmetry_factor and cam_state[4] (angle) determine the direction and magnitude
            horizontal_impulse = -np.sign(np.sin(cam_state[4])) * vibration_amplitude * vibration_frequency * asymmetry_factor * cam_elasticity * 0.01 # Simplified model
            cam_state[2] += horizontal_impulse / CAM_MASS

            # Apply friction to angular velocity (rough model)
            cam_state[5] *= 0.95 # Damping

        # Update position and angular position
        cam_state[0] += cam_state[2] * DT
        cam_state[1] += cam_state[3] * DT
        cam_state[4] += cam_state[5] * DT

        history.append(np.copy(cam_state))

    return np.array(history)

def plot_animation(history, vibration_frequency, vibration_amplitude, asymmetry_factor):
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.set_xlim(-2, 2)
    ax.set_ylim(PLATE_HEIGHT - vibration_amplitude - CAM_RADIUS - 0.5, CAM_RADIUS + 0.5)
    ax.set_aspect('equal')
    ax.grid(True)

    # Plate
    plate_line, = ax.plot([], [], 'k-', lw=2)

    # Cam
    cam_circle = plt.Circle((0,0), CAM_RADIUS, color='blue', fill=True, ec='black')
    ax.add_patch(cam_circle)

    # Cam asymmetry indicator
    cam_asymmetry_line, = ax.plot([], [], 'r-', lw=2)

    ax.set_title(f'Vibration-to-Motion Siphon (Freq: {vibration_frequency} Hz, Amp: {vibration_amplitude})')
    ax.set_xlabel('Horizontal Position')
    ax.set_ylabel('Vertical Position')

    def init():
        plate_line.set_data([], [])
        cam_circle.set_center((0,0))
        cam_asymmetry_line.set_data([],[])
        return plate_line, cam_circle, cam_asymmetry_line

    def update(frame):
        t = frame * DT
        plate_y = PLATE_HEIGHT + vibration_amplitude * np.sin(2 * np.pi * vibration_frequency * t)
        plate_line.set_data([-5, 5], [plate_y, plate_y])

        cam_circle.set_center((history[frame, 0], history[frame, 1]))

        # Update asymmetry indicator
        effective_radius = CAM_RADIUS + asymmetry_factor * np.sin(history[frame, 4])
        x_asym = history[frame, 0] + effective_radius * np.sin(history[frame, 4])
        y_asym = history[frame, 1] - effective_radius * np.cos(history[frame, 4])
        cam_asymmetry_line.set_data([history[frame, 0], x_asym], [history[frame, 1], y_asym])

        ax.set_xlim(history[frame, 0] - 2, history[frame, 0] + 2) # Follow the cam horizontally
        return plate_line, cam_circle, cam_asymmetry_line

    ani = FuncAnimation(fig, update, frames=len(history), init_func=init, blit=True, interval=DT*1000)
    plt.close(fig) # Prevent static plot from showing
    return ani

# --- Interactive Widgets ---
vibration_freq_slider = widgets.FloatSlider(value=50.0, min=10.0, max=150.0, step=5.0,
                                          description='Vibration Freq (Hz):', continuous_update=False)
vibration_amp_slider = widgets.FloatSlider(value=0.1, min=0.01, max=0.5, step=0.01,
                                         description='Vibration Amp:', continuous_update=False)
cam_elasticity_slider = widgets.FloatSlider(value=0.5, min=0.1, max=0.9, step=0.1,
                                           description='Bounce-Gap (Elasticity):', continuous_update=False)
asymmetry_factor_slider = widgets.FloatSlider(value=0.3, min=0.0, max=0.5, step=0.05,
                                              description='Asymmetry Factor:', continuous_update=False)


def interactive_siphon_analogue(vibration_frequency, vibration_amplitude, cam_elasticity, asymmetry_factor):
    history = simulate_cam_roll(vibration_frequency, vibration_amplitude, cam_elasticity, asymmetry_factor)
    ani = plot_animation(history, vibration_frequency, vibration_amplitude, asymmetry_factor)
    display(HTML(ani.to_jshtml()))

# Display the interactive widget
print("Adjust the sliders to observe how vibration is converted to directed motion!")
display(widgets.interactive(interactive_siphon_analogue,
                            vibration_frequency=vibration_freq_slider,
                            vibration_amplitude=vibration_amp_slider,
                            cam_elasticity=cam_elasticity_slider,
                            asymmetry_factor=asymmetry_factor_slider))
```
